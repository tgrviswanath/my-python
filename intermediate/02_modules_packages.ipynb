## 1. Inspecting `sys.modules` and `sys.path`

```python
import sys

print('=== sys.path (module search locations) ===')
for i, p in enumerate(sys.path):
    print(f'  [{i}] {p}')

print()
print(f'Total modules in sys.modules: {len(sys.modules)}')
print('Sample cached modules:', list(sys.modules.keys())[:8])
```

## 2. Import Timing  Cached vs Fresh

```python
import time
import sys

# Remove json from cache to simulate fresh import
if 'json' in sys.modules:
    del sys.modules['json']

t0 = time.perf_counter()
import json
t1 = time.perf_counter()
fresh_time = (t1 - t0) * 1000

t0 = time.perf_counter()
import json  # cached
t1 = time.perf_counter()
cached_time = (t1 - t0) * 1000

print(f'Fresh import:  {fresh_time:.4f} ms')
print(f'Cached import: {cached_time:.6f} ms')
print(f'Speedup: {fresh_time/max(cached_time, 0.0001):.0f}x')
```

## 3. Dynamic Import with importlib

```python
import importlib
import importlib.util

# Dynamically import modules by name
modules_to_check = ['json', 'csv', 'pathlib', 'collections', 'numpy', 'pandas']

print('Module availability check:')
for mod_name in modules_to_check:
    spec = importlib.util.find_spec(mod_name)
    status = 'FOUND' if spec else 'NOT INSTALLED'
    location = spec.origin if spec else 'N/A'
    print(f'  {mod_name:<15} {status:<15} {location}')

print()
# Dynamic import
json_mod = importlib.import_module('json')
result = json_mod.dumps({'dynamic': True, 'value': 42})
print('Dynamic json.dumps:', result)
```

## 4. `__name__ == '__main__'` Demo

```python
# Simulate what happens when a module is imported vs run directly
import tempfile, os, sys, importlib

module_code = '''
def greet(name):
    return f"Hello, {name}!"

print(f"Module __name__ is: {__name__}")

if __name__ == "__main__":
    print("Running as script!")
    print(greet("World"))
else:
    print("Imported as module  main block skipped")
'''

# Write to a temp file and import it
with tempfile.NamedTemporaryFile(mode='w', suffix='.py', delete=False, dir='.') as f:
    f.write(module_code)
    tmp_path = f.name

mod_name = os.path.basename(tmp_path)[:-3]
sys.path.insert(0, '.')

print('=== Importing the module ===')
mod = importlib.import_module(mod_name)
print('greet result:', mod.greet('Python'))

# Cleanup
os.unlink(tmp_path)
del sys.modules[mod_name]
sys.path.pop(0)
```

## 5. `collections` Module Deep Dive

```python
from collections import defaultdict, Counter, deque, namedtuple, ChainMap, OrderedDict

# --- defaultdict ---
print('=== defaultdict ===')
graph = defaultdict(list)
edges = [(1, 2), (1, 3), (2, 4), (3, 4), (4, 5)]
for u, v in edges:
    graph[u].append(v)
    graph[v].append(u)
print('Graph adjacency:', dict(graph))

# --- Counter ---
print()
print('=== Counter ===')
text = 'the quick brown fox jumps over the lazy dog the fox'
word_freq = Counter(text.split())
print('Top 5 words:', word_freq.most_common(5))
print('Total words:', sum(word_freq.values()))
print('Unique words:', len(word_freq))

# Counter arithmetic
c1 = Counter(a=3, b=2, c=1)
c2 = Counter(a=1, b=2, c=3)
print('c1 + c2:', c1 + c2)
print('c1 - c2:', c1 - c2)

# --- deque ---
print()
print('=== deque (sliding window) ===')
def moving_average(data, window):
    dq = deque(maxlen=window)
    result = []
    for val in data:
        dq.append(val)
        result.append(sum(dq) / len(dq))
    return result

data = [1, 3, 5, 7, 9, 11, 13]
print('Data:', data)
print('3-period MA:', [f'{v:.2f}' for v in moving_average(data, 3)])

# --- namedtuple ---
print()
print('=== namedtuple ===')
Point3D = namedtuple('Point3D', ['x', 'y', 'z'])
p = Point3D(1, 2, 3)
print(f'Point: {p}')
print(f'x={p.x}, y={p.y}, z={p.z}')
print(f'As dict: {p._asdict()}')
p2 = p._replace(z=10)
print(f'Replaced z: {p2}')
```

## 6. `itertools`  Iterator Building Blocks

```python
import itertools

# chain
print('chain:', list(itertools.chain([1,2], [3,4], [5])))

# islice
print('islice(range(20), 5, 15, 2):', list(itertools.islice(range(20), 5, 15, 2)))

# product
print('product(AB, 12):', list(itertools.product('AB', '12')))

# combinations
print('combinations(ABCD, 2):', list(itertools.combinations('ABCD', 2)))

# permutations
print('permutations(ABC, 2):', list(itertools.permutations('ABC', 2)))

# groupby
print()
print('=== groupby ===')
employees = [
    ('Alice', 'Engineering'), ('Bob', 'Engineering'),
    ('Carol', 'Marketing'), ('Dave', 'Marketing'),
    ('Eve', 'HR'),
]
employees.sort(key=lambda x: x[1])  # must sort before groupby
for dept, group in itertools.groupby(employees, key=lambda x: x[1]):
    names = [e[0] for e in group]
    print(f'  {dept}: {names}')

# accumulate
print()
import operator
data = [1, 2, 3, 4, 5]
print('accumulate (sum):', list(itertools.accumulate(data)))
print('accumulate (product):', list(itertools.accumulate(data, operator.mul)))
print('accumulate (max):', list(itertools.accumulate(data, max)))

# takewhile / dropwhile
print()
nums = [1, 3, 5, 2, 4, 6, 7]
print('takewhile(<5):', list(itertools.takewhile(lambda x: x < 5, nums)))
print('dropwhile(<5):', list(itertools.dropwhile(lambda x: x < 5, nums)))
```

## 7. `functools`  Higher-Order Functions

```python
import functools
import time

# lru_cache
print('=== lru_cache ===')
@functools.lru_cache(maxsize=None)
def fib(n):
    if n < 2: return n
    return fib(n-1) + fib(n-2)

t0 = time.perf_counter()
result = fib(35)
t1 = time.perf_counter()
print(f'fib(35) = {result} in {(t1-t0)*1000:.3f} ms')
print(f'Cache info: {fib.cache_info()}')

# partial
print()
print('=== partial ===')
def power(base, exp):
    return base ** exp

square = functools.partial(power, exp=2)
cube   = functools.partial(power, exp=3)
print('square(5):', square(5))
print('cube(3):', cube(3))

# reduce
print()
print('=== reduce ===')
from functools import reduce
factorial = lambda n: reduce(lambda a, b: a * b, range(1, n+1), 1)
print('10! =', factorial(10))

# total_ordering
print()
print('=== total_ordering ===')
from functools import total_ordering

@total_ordering
class Version:
    def __init__(self, major, minor, patch):
        self.v = (major, minor, patch)
    def __eq__(self, other): return self.v == other.v
    def __lt__(self, other): return self.v < other.v
    def __repr__(self): return f'v{".".join(map(str, self.v))}'

versions = [Version(2,0,0), Version(1,9,5), Version(2,1,0), Version(1,0,0)]
print('Sorted:', sorted(versions))
print('Max:', max(versions))
```

## 8. `pathlib`  Modern File System Operations

```python
from pathlib import Path
import tempfile

# Create a temp directory to work in
with tempfile.TemporaryDirectory() as tmpdir:
    base = Path(tmpdir)

    # Create directory structure
    (base / 'src' / 'utils').mkdir(parents=True)
    (base / 'tests').mkdir()
    (base / 'data').mkdir()

    # Write files
    (base / 'src' / 'main.py').write_text('print("hello")')
    (base / 'src' / 'utils' / 'helpers.py').write_text('def add(a,b): return a+b')
    (base / 'tests' / 'test_main.py').write_text('import pytest')
    (base / 'README.md').write_text('# My Project')

    # Path operations
    p = base / 'src' / 'main.py'
    print(f'name:   {p.name}')
    print(f'stem:   {p.stem}')
    print(f'suffix: {p.suffix}')
    print(f'parent: {p.parent.name}')
    print(f'exists: {p.exists()}')
    print(f'is_file: {p.is_file()}')

    # Glob
    print()
    print('All .py files:')
    for f in sorted(base.glob('**/*.py')):
        print(f'  {f.relative_to(base)}')

    # Read/write
    content = (base / 'README.md').read_text()
    print()
    print('README content:', content)

    # Rename
    old = base / 'README.md'
    new = old.with_name('README.txt')
    old.rename(new)
    print('Renamed to:', new.name)

print('Temp directory cleaned up.')
```

## 9. `contextlib` Utilities

```python
from contextlib import contextmanager, suppress, redirect_stdout, ExitStack
import io

# contextmanager decorator
@contextmanager
def managed_resource(name):
    print(f'Acquiring {name}')
    try:
        yield name.upper()
    except Exception as e:
        print(f'Error in {name}: {e}')
        raise
    finally:
        print(f'Releasing {name}')

with managed_resource('database') as res:
    print(f'Using resource: {res}')

print()

# suppress
print('=== suppress ===')
with suppress(FileNotFoundError, PermissionError):
    open('/nonexistent/path/file.txt')
print('No exception raised (suppressed)')

# redirect_stdout
print()
print('=== redirect_stdout ===')
buffer = io.StringIO()
with redirect_stdout(buffer):
    print('This goes to the buffer')
    print('So does this')
captured = buffer.getvalue()
print('Captured output:')
print(repr(captured))

# ExitStack  dynamic context managers
print()
print('=== ExitStack ===')
resources = ['db', 'cache', 'queue']
with ExitStack() as stack:
    opened = [stack.enter_context(managed_resource(r)) for r in resources]
    print('All resources open:', opened)
```
