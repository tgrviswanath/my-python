## 1. Class Anatomy: `__dict__` vs `__slots__`

```python
import sys

class WithDict:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

class WithSlots:
    __slots__ = ('x', 'y', 'z')
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

d = WithDict(1, 2, 3)
s = WithSlots(1, 2, 3)

print(f'WithDict  instance size : {sys.getsizeof(d)} bytes')
print(f'WithDict  __dict__ size : {sys.getsizeof(d.__dict__)} bytes')
print(f'WithSlots instance size : {sys.getsizeof(s)} bytes')
print(f'WithSlots has __dict__  : {hasattr(s, "__dict__")}')
print()
print('WithDict  __dict__:', d.__dict__)
print('WithSlots __slots__:', WithSlots.__slots__)
```

## 2. Memory Benchmark: 1 million objects

```python
import tracemalloc

N = 100_000

tracemalloc.start()
objs_dict = [WithDict(i, i*2, i*3) for i in range(N)]
snap1 = tracemalloc.take_snapshot()
tracemalloc.stop()

tracemalloc.start()
objs_slots = [WithSlots(i, i*2, i*3) for i in range(N)]
snap2 = tracemalloc.take_snapshot()
tracemalloc.stop()

stats1 = snap1.statistics('lineno')
stats2 = snap2.statistics('lineno')

total1 = sum(s.size for s in stats1) / 1024 / 1024
total2 = sum(s.size for s in stats2) / 1024 / 1024

print(f'WithDict  ({N:,} objects): {total1:.2f} MB')
print(f'WithSlots ({N:,} objects): {total2:.2f} MB')
print(f'Savings: {(1 - total2/total1)*100:.1f}%')
```

## 3. MRO Visualization  Diamond Problem

```python
class A:
    def greet(self):
        print('A.greet')
        # no super() call here  A is the top

class B(A):
    def greet(self):
        print('B.greet')
        super().greet()

class C(A):
    def greet(self):
        print('C.greet')
        super().greet()

class D(B, C):
    def greet(self):
        print('D.greet')
        super().greet()

print('MRO for D:')
for i, cls in enumerate(D.__mro__):
    print(f'  {i}: {cls.__name__}')

print()
print('Calling D().greet():')
D().greet()
```

## 4. Dunder Methods  Full Vector Class

```python
import math

class Vector:
    """2D vector with full dunder method support."""

    def __init__(self, x, y):
        self.x = float(x)
        self.y = float(y)

    # --- representation ---
    def __repr__(self):
        return f'Vector({self.x!r}, {self.y!r})'

    def __str__(self):
        return f'({self.x}, {self.y})'

    # --- arithmetic ---
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __rmul__(self, scalar):
        return self.__mul__(scalar)

    def __neg__(self):
        return Vector(-self.x, -self.y)

    def __abs__(self):
        return math.hypot(self.x, self.y)

    # --- comparison ---
    def __eq__(self, other):
        if not isinstance(other, Vector):
            return NotImplemented
        return math.isclose(self.x, other.x) and math.isclose(self.y, other.y)

    def __hash__(self):
        return hash((self.x, self.y))

    # --- container-like ---
    def __len__(self):
        return 2

    def __getitem__(self, idx):
        return (self.x, self.y)[idx]

    def __iter__(self):
        yield self.x
        yield self.y

    # --- callable ---
    def __call__(self, scale):
        return self * scale

    # --- bool ---
    def __bool__(self):
        return abs(self) != 0

v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f'repr: {repr(v1)}')
print(f'str:  {str(v1)}')
print(f'v1 + v2 = {v1 + v2}')
print(f'3 * v1  = {3 * v1}')
print(f'abs(v1) = {abs(v1)}')
print(f'len(v1) = {len(v1)}')
print(f'list(v1) = {list(v1)}')
print(f'v1(2) = {v1(2)}')
print(f'v1 in set: {v1 in {v1, v2}}')
x, y = v1
print(f'unpacked: x={x}, y={y}')
```

## 5. Context Manager  `__enter__` / `__exit__`

```python
import time

class Timer:
    """Context manager that measures elapsed time."""

    def __enter__(self):
        self._start = time.perf_counter()
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.elapsed = time.perf_counter() - self._start
        print(f'Elapsed: {self.elapsed*1000:.3f} ms')
        return False  # don't suppress exceptions

with Timer() as t:
    total = sum(range(1_000_000))

print(f'Sum: {total}')
print(f'Stored elapsed: {t.elapsed*1000:.3f} ms')
```

## 6. Iterator Protocol  Fibonacci

```python
class Fibonacci:
    """Infinite Fibonacci iterator."""

    def __init__(self):
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        return value

import itertools
fib = Fibonacci()
first_10 = list(itertools.islice(fib, 10))
print('First 10 Fibonacci:', first_10)
```

## 7. @property with Validation

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius  # uses the setter

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if not isinstance(value, (int, float)):
            raise TypeError(f'radius must be numeric, got {type(value).__name__}')
        if value < 0:
            raise ValueError(f'radius must be >= 0, got {value}')
        self._radius = float(value)

    @property
    def area(self):
        return math.pi * self._radius ** 2

    @property
    def circumference(self):
        return 2 * math.pi * self._radius

    def __repr__(self):
        return f'Circle(radius={self._radius})'

c = Circle(5)
print(c)
print(f'Area: {c.area:.4f}')
print(f'Circumference: {c.circumference:.4f}')

c.radius = 10
print(f'After resize: {c}')

try:
    c.radius = -1
except ValueError as e:
    print(f'Error: {e}')
```

## 8. Metaclass  Auto-registering Plugins

```python
class PluginMeta(type):
    registry = {}

    def __new__(mcs, name, bases, namespace):
        cls = super().__new__(mcs, name, bases, namespace)
        if bases:  # skip the base class itself
            plugin_name = namespace.get('name', name.lower())
            mcs.registry[plugin_name] = cls
            print(f'  Registered plugin: {plugin_name!r} -> {name}')
        return cls

class BasePlugin(metaclass=PluginMeta):
    def run(self):
        raise NotImplementedError

print('Defining plugins...')

class CSVPlugin(BasePlugin):
    name = 'csv'
    def run(self): return 'Processing CSV'

class JSONPlugin(BasePlugin):
    name = 'json'
    def run(self): return 'Processing JSON'

class XMLPlugin(BasePlugin):
    name = 'xml'
    def run(self): return 'Processing XML'

print()
print('Registry:', list(PluginMeta.registry.keys()))
print()

# Use the registry
for name, cls in PluginMeta.registry.items():
    result = cls().run()
    print(f'{name}: {result}')
```

## 9. Abstract Base Classes

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self) -> float: ...

    @abstractmethod
    def perimeter(self) -> float: ...

    def describe(self):
        return (f'{self.__class__.__name__}: '
                f'area={self.area():.2f}, '
                f'perimeter={self.perimeter():.2f}')

class Rectangle(Shape):
    def __init__(self, w, h): self.w, self.h = w, h
    def area(self): return self.w * self.h
    def perimeter(self): return 2 * (self.w + self.h)

class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return math.pi * self.r ** 2
    def perimeter(self): return 2 * math.pi * self.r

shapes = [Rectangle(3, 4), Circle(5), Rectangle(10, 2)]
for s in shapes:
    print(s.describe())

print()
print('Trying to instantiate Shape directly...')
try:
    Shape()
except TypeError as e:
    print(f'TypeError: {e}')
```

## 10. Dataclasses  Full Example

```python
from dataclasses import dataclass, field
from typing import ClassVar

@dataclass(order=True)
class Employee:
    sort_index: float = field(init=False, repr=False)
    name: str
    department: str
    salary: float
    skills: list = field(default_factory=list)
    _count: ClassVar[int] = 0

    def __post_init__(self):
        self.sort_index = self.salary
        Employee._count += 1

    def give_raise(self, pct: float):
        self.salary *= (1 + pct / 100)
        self.sort_index = self.salary

employees = [
    Employee('Alice', 'Engineering', 95_000, ['Python', 'ML']),
    Employee('Bob', 'Marketing', 72_000, ['Excel', 'Tableau']),
    Employee('Carol', 'Engineering', 105_000, ['Python', 'Go']),
]

print('All employees:')
for e in employees:
    print(f'  {e}')

print()
print('Sorted by salary:')
for e in sorted(employees):
    print(f'  {e.name}: ${e.salary:,.0f}')

print()
print(f'Total employees created: {Employee._count}')

employees[0].give_raise(10)
print(f'Alice after 10% raise: ${employees[0].salary:,.0f}')
```

## 11. Mixin Pattern

```python
import json

class JSONMixin:
    def to_json(self):
        return json.dumps(self.__dict__, default=str, indent=2)

    @classmethod
    def from_json(cls, json_str):
        data = json.loads(json_str)
        obj = cls.__new__(cls)
        obj.__dict__.update(data)
        return obj

class ValidateMixin:
    def validate(self):
        errors = []
        for attr, expected_type in getattr(self, '_types', {}).items():
            val = getattr(self, attr, None)
            if not isinstance(val, expected_type):
                errors.append(f'{attr}: expected {expected_type.__name__}, got {type(val).__name__}')
        return errors

class Product(JSONMixin, ValidateMixin):
    _types = {'name': str, 'price': float, 'stock': int}

    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

p = Product('Widget', 9.99, 100)
print('JSON:')
print(p.to_json())

print()
print('Validation errors:', p.validate())

bad = Product('Gadget', 'free', 'many')  # wrong types
print('Bad product errors:', bad.validate())

# Round-trip
p2 = Product.from_json(p.to_json())
print()
print('Round-trip:', p2.name, p2.price)
```

## 12. `__new__`  Singleton & Immutable Subclass

```python
# Singleton via __new__
class Singleton:
    _instance = None

    def __new__(cls, *args, **kwargs):
        if cls._instance is None:
            print(f'Creating new {cls.__name__} instance')
            cls._instance = super().__new__(cls)
        else:
            print(f'Returning existing {cls.__name__} instance')
        return cls._instance

    def __init__(self, value):
        self.value = value

a = Singleton(42)
b = Singleton(99)
print(f'a is b: {a is b}')
print(f'a.value: {a.value}')  # 99  __init__ ran again

print()

# Immutable int subclass via __new__
class EvenInt(int):
    def __new__(cls, value):
        if value % 2 != 0:
            raise ValueError(f'{value} is not even')
        return super().__new__(cls, value)

e = EvenInt(4)
print(f'EvenInt(4) = {e}, type = {type(e).__name__}')
print(f'e + 2 = {e + 2}')

try:
    EvenInt(3)
except ValueError as err:
    print(f'Error: {err}')
```
