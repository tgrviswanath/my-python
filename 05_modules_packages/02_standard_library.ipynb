## 1. collections — Counter

```python
from collections import Counter

# Count word frequencies
words = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
c = Counter(words)
print('Counter:', c)
print('most_common(2):', c.most_common(2))
print('Missing key returns 0:', c['missing'])  # no KeyError!

# Counter arithmetic
c1 = Counter(a=3, b=2, c=1)
c2 = Counter(a=1, b=4)
print('\nAddition:', c1 + c2)
print('Subtraction:', c1 - c2)
print('Intersection (min):', c1 & c2)
print('Union (max):', c1 | c2)

# Count characters
char_count = Counter('mississippi')
print('\nCharacter count:', char_count.most_common(3))
```

## 2. collections — defaultdict

```python
from collections import defaultdict

# Group words by first letter
words = ['apple', 'ant', 'banana', 'bear', 'cherry', 'avocado']
grouped = defaultdict(list)
for word in words:
    grouped[word[0]].append(word)
print('Grouped by first letter:', dict(grouped))

# Count with defaultdict(int)
word_count = defaultdict(int)
for word in words:
    word_count[word] += 1
print('\nWord count:', dict(word_count))

# Nested defaultdict
matrix = defaultdict(lambda: defaultdict(int))
matrix['row1']['col1'] = 10
matrix['row1']['col2'] = 20
matrix['row2']['col1'] = 30
print('\nMatrix row1:', dict(matrix['row1']))
print('Missing value:', matrix['row3']['col1'])  # 0, no error
```

## 3. collections — deque, namedtuple, OrderedDict, ChainMap

```python
from collections import deque, namedtuple, OrderedDict, ChainMap

# deque — O(1) operations at both ends
dq = deque([1, 2, 3, 4, 5])
dq.appendleft(0)
dq.append(6)
print('deque after appends:', list(dq))
print('popleft:', dq.popleft())
print('pop:', dq.pop())

# Sliding window with maxlen
window = deque(maxlen=3)
print('\nSliding window (maxlen=3):')
for i in range(6):
    window.append(i)
    print(f'  After append({i}):', list(window))

# namedtuple
Point = namedtuple('Point', ['x', 'y'])
p = Point(3, 4)
print('\nPoint:', p)
print('p.x:', p.x, '| p[0]:', p[0])
print('_asdict:', p._asdict())
print('_replace:', p._replace(x=10))

# ChainMap — layered config
defaults = {'color': 'red', 'size': 'medium', 'debug': False}
user_prefs = {'color': 'blue'}
config = ChainMap(user_prefs, defaults)
print('\nChainMap color:', config['color'])  # from user_prefs
print('ChainMap size:', config['size'])    # from defaults
```

## 4. itertools — chain, islice, product, combinations, permutations

```python
import itertools

# chain — flatten iterables
result = list(itertools.chain([1, 2], [3, 4], [5, 6]))
print('chain:', result)

nested = [[1, 2], [3, 4], [5, 6]]
flat = list(itertools.chain.from_iterable(nested))
print('chain.from_iterable:', flat)

# islice — lazy slicing (memory efficient)
gen = (x**2 for x in range(1_000_000))
first_five = list(itertools.islice(gen, 5))
print('\nislice first 5 squares:', first_five)

# product — Cartesian product
colors = ['red', 'blue']
sizes = ['S', 'M', 'L']
variants = list(itertools.product(colors, sizes))
print('\nproduct:', variants)

# combinations — order doesn't matter, no repetition
combos = list(itertools.combinations([1, 2, 3, 4], 2))
print('\ncombinations(4,2):', combos)
print('Count:', len(combos))  # C(4,2) = 6

# permutations — order matters
perms = list(itertools.permutations([1, 2, 3], 2))
print('\npermutations(3,2):', perms)
print('Count:', len(perms))  # P(3,2) = 6
```

## 5. itertools — groupby, accumulate, cycle, repeat

```python
import itertools
import operator

# groupby — MUST be sorted first!
data = [('Alice', 'Engineering'), ('Bob', 'Engineering'), 
        ('Charlie', 'Marketing'), ('Diana', 'Marketing'),
        ('Eve', 'Engineering')]

# Sort by department first
sorted_data = sorted(data, key=lambda x: x[1])
print('groupby (sorted):')
for dept, employees in itertools.groupby(sorted_data, key=lambda x: x[1]):
    print(f'  {dept}: {[e[0] for e in employees]}')

# accumulate — running totals
nums = [1, 2, 3, 4, 5]
running_sum = list(itertools.accumulate(nums))
running_product = list(itertools.accumulate(nums, operator.mul))
running_max = list(itertools.accumulate(nums, max))
print('\nrunning sum:', running_sum)
print('running product:', running_product)
print('running max:', running_max)

# cycle — infinite cycling (use islice to limit!)
colors = itertools.cycle(['red', 'green', 'blue'])
print('\ncycle (first 7):', [next(colors) for _ in range(7)])

# repeat
print('repeat:', list(itertools.repeat(42, 4)))
```

## 6. functools — lru_cache, partial, reduce, wraps

```python
import functools
import time

# lru_cache — memoization
@functools.lru_cache(maxsize=128)
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

start = time.perf_counter()
result = fibonacci(50)
elapsed = time.perf_counter() - start
print(f'fibonacci(50) = {result}')
print(f'Time: {elapsed:.6f}s')
print('Cache info:', fibonacci.cache_info())

# partial — pre-fill arguments
def power(base, exponent):
    return base ** exponent

square = functools.partial(power, exponent=2)
cube = functools.partial(power, exponent=3)
print('\nsquare(5):', square(5))
print('cube(3):', cube(3))

# reduce — fold left
from functools import reduce
product = reduce(lambda acc, x: acc * x, [1, 2, 3, 4, 5])
print('\nreduce product:', product)

# wraps — preserve function metadata
def timer(func):
    @functools.wraps(func)  # preserves __name__, __doc__
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f'{func.__name__} took {elapsed:.6f}s')
        return result
    return wrapper

@timer
def slow_function(n):
    """A slow function for demonstration."""
    return sum(range(n))

result = slow_function(1_000_000)
print('Function name preserved:', slow_function.__name__)  # 'slow_function', not 'wrapper'
print('Docstring preserved:', slow_function.__doc__)
```

## 7. pathlib — Modern Path Operations

```python
from pathlib import Path
import tempfile

# Path properties
p = Path('/home/user/documents/report.pdf')
print('name:', p.name)       # report.pdf
print('stem:', p.stem)       # report
print('suffix:', p.suffix)   # .pdf
print('parent:', p.parent)   # /home/user/documents
print('parts:', p.parts)

# Join paths with /
config = Path.home() / '.config' / 'myapp' / 'settings.json'
print('\nConfig path:', config)

# Current directory
cwd = Path.cwd()
print('CWD:', cwd)

# Read/write in temp directory
with tempfile.TemporaryDirectory() as tmpdir:
    tmp = Path(tmpdir)
    
    # Write and read text
    txt_file = tmp / 'hello.txt'
    txt_file.write_text('Hello, pathlib!')
    content = txt_file.read_text()
    print('\nRead:', content)
    
    # Create nested directories
    nested = tmp / 'a' / 'b' / 'c'
    nested.mkdir(parents=True, exist_ok=True)
    print('Created nested dir:', nested.exists())
    
    # Glob
    (tmp / 'file1.py').write_text('# py file 1')
    (tmp / 'file2.py').write_text('# py file 2')
    (tmp / 'data.txt').write_text('data')
    
    py_files = list(tmp.glob('*.py'))
    print('Python files:', [f.name for f in py_files])
```

## 8. datetime — Dates, Times, Timezones

```python
from datetime import date, time, datetime, timedelta, timezone

# date
today = date.today()
d = date(2024, 1, 15)
print('date:', d)
print('strftime:', d.strftime('%B %d, %Y'))
print('isoformat:', d.isoformat())

# datetime
now = datetime.now()
utc_now = datetime.now(timezone.utc)
print('\nnow:', now)
print('utc_now:', utc_now)

# Formatting and parsing
formatted = now.strftime('%Y-%m-%d %H:%M:%S')
parsed = datetime.strptime('2024-01-15 14:30:45', '%Y-%m-%d %H:%M:%S')
print('formatted:', formatted)
print('parsed:', parsed)

# timedelta arithmetic
delta = timedelta(days=7, hours=3)
future = now + delta
past = now - timedelta(days=30)
diff = future - now
print('\n7 days 3 hours from now:', future.strftime('%Y-%m-%d %H:%M'))
print('30 days ago:', past.strftime('%Y-%m-%d'))
print('diff.days:', diff.days)

# Timezone-aware (CORRECT way)
utc = timezone.utc
eastern = timezone(timedelta(hours=-5))
aware = datetime(2024, 6, 15, 12, 0, tzinfo=utc)
eastern_time = aware.astimezone(eastern)
print('\nUTC:', aware.isoformat())
print('Eastern:', eastern_time.isoformat())
```

## 9. json — Serialization

```python
import json
from datetime import datetime

# Basic dumps/loads
data = {'name': 'Alice', 'age': 30, 'scores': [95, 87, 92], 'active': True, 'address': None}
json_str = json.dumps(data)
print('JSON string:', json_str)

# Pretty print
pretty = json.dumps(data, indent=2, sort_keys=True)
print('\nPretty JSON:')
print(pretty)

# loads
parsed = json.loads(json_str)
print('\nParsed name:', parsed['name'])
print('Type:', type(parsed))

# Custom serializer for non-serializable types
def json_serializer(obj):
    if isinstance(obj, datetime):
        return obj.isoformat()
    if isinstance(obj, set):
        return list(obj)
    raise TypeError(f'Object of type {type(obj).__name__} is not JSON serializable')

data_with_special = {
    'timestamp': datetime.now(),
    'tags': {'python', 'json', 'interview'},
    'value': 42
}
result = json.dumps(data_with_special, default=json_serializer, indent=2)
print('\nCustom serializer result:')
print(result)
```

## 10. re — Regular Expressions

```python
import re

text = 'Contact alice@example.com or bob@test.org. Phone: 555-1234 or (800) 555-5678'

# match vs search vs findall
print('match (start only):', re.match(r'Contact', text))  # matches
print('match (not start):', re.match(r'alice', text))     # None!
print('search (anywhere):', re.search(r'alice', text).group())

# findall
emails = re.findall(r'\b[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}\b', text)
print('\nEmails found:', emails)

phones = re.findall(r'\(?\d{3}\)?[\s.-]?\d{3}[\s.-]?\d{4}', text)
print('Phones found:', phones)

# sub — replace
redacted = re.sub(r'\b[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}\b', '[EMAIL]', text)
print('\nRedacted:', redacted)

# Groups
date_pattern = re.compile(r'(\d{4})-(\d{2})-(\d{2})')
m = date_pattern.search('Date: 2024-01-15')
if m:
    print('\nGroups:', m.groups())
    print('Group 1 (year):', m.group(1))

# Named groups
named = re.compile(r'(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})')
m = named.search('Date: 2024-01-15')
if m:
    print('Named groups:', m.groupdict())
    print('Year:', m.group('year'))
```

## 11. typing — Type Hints

```python
from typing import Optional, Union, List, Dict, Callable, TypeVar, Generic, Protocol, Literal, TypedDict

# Optional — value or None
def find_user(user_id: int) -> Optional[str]:
    users = {1: 'Alice', 2: 'Bob'}
    return users.get(user_id)

print('find_user(1):', find_user(1))
print('find_user(99):', find_user(99))

# Union
def process(value: Union[int, str]) -> str:
    return str(value).upper()

# TypeVar + Generic
T = TypeVar('T')

class Stack(Generic[T]):
    def __init__(self) -> None:
        self._items: List[T] = []
    
    def push(self, item: T) -> None:
        self._items.append(item)
    
    def pop(self) -> T:
        if not self._items:
            raise IndexError('Stack is empty')
        return self._items.pop()
    
    def __len__(self) -> int:
        return len(self._items)

s: Stack[int] = Stack()
s.push(1)
s.push(2)
s.push(3)
print('Stack pop:', s.pop())
print('Stack len:', len(s))

# Protocol — structural subtyping
class Drawable(Protocol):
    def draw(self) -> str: ...

class Circle:
    def draw(self) -> str:
        return 'Drawing circle'

class Square:
    def draw(self) -> str:
        return 'Drawing square'

def render(shape: Drawable) -> None:
    print(shape.draw())

render(Circle())  # works — no inheritance needed!
render(Square())

# TypedDict
class UserInfo(TypedDict):
    name: str
    age: int
    email: str

user: UserInfo = {'name': 'Alice', 'age': 30, 'email': 'alice@example.com'}
print('\nTypedDict user:', user['name'])

# Literal
def set_log_level(level: Literal['DEBUG', 'INFO', 'WARNING', 'ERROR']) -> None:
    print(f'Log level set to: {level}')

set_log_level('INFO')
```
