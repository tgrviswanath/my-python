# Python Dictionaries — Interview Preparation

Covers: creation, CRUD, methods, comprehensions, defaultdict, OrderedDict, Counter, ChainMap, merging, iteration, and interview Q&A.

## 1. Creation and CRUD

```python
# Multiple ways to create dicts
d1 = {'name': 'Alice', 'age': 30}
d2 = dict(name='Alice', age=30)
d3 = dict([('name', 'Alice'), ('age', 30)])
d4 = dict(zip(['a', 'b', 'c'], [1, 2, 3]))

print('d1:', d1)
print('d4:', d4)

# CRUD
d = {'name': 'Alice', 'age': 30}
d['city'] = 'NYC'          # Create
print('Read name:', d['name'])  # Read
d['age'] = 31              # Update
del d['city']              # Delete
print('After CRUD:', d)
```

```python
# get() vs [] — key difference
d = {'a': 1, 'b': 2}

print('d["a"]:', d['a'])           # 1
# d['missing']                     # KeyError!

print('d.get("a"):', d.get('a'))   # 1
print('d.get("x"):', d.get('x'))   # None — no error
print('d.get("x", 0):', d.get('x', 0))  # 0 — default value

# Use [] when missing key is a bug (fail fast)
# Use .get() when missing key is expected
```

## 2. Dict Methods

```python
d = {'a': 1, 'b': 2, 'c': 3}

# keys(), values(), items() — return VIEW objects (dynamic!)
keys = d.keys()
print('keys:', keys)

d['d'] = 4  # modify dict
print('keys after adding d:', keys)  # view reflects the change!

# Iterate
for k, v in d.items():
    print(f'  {k}: {v}')
```

```python
# setdefault — get or set default
d = {'a': 1}

# Returns existing value if key exists
print('setdefault a:', d.setdefault('a', 99))  # 1 — not changed

# Sets and returns default if key missing
print('setdefault b:', d.setdefault('b', 99))  # 99 — added
print('d after setdefault:', d)

# Classic use: grouping
groups = {}
for item in ['apple', 'banana', 'avocado', 'blueberry', 'cherry']:
    groups.setdefault(item[0], []).append(item)
print('groups:', groups)
```

```python
# pop() and popitem()
d = {'a': 1, 'b': 2, 'c': 3}

val = d.pop('a')           # returns 1, removes 'a'
print('pop(a):', val, '  d:', d)

val = d.pop('x', None)     # returns None, no KeyError
print('pop(x, None):', val)

item = d.popitem()         # removes and returns last inserted (Python 3.7+)
print('popitem():', item, '  d:', d)
```

## 3. Dict Comprehension

```python
# Basic
squares = {x: x**2 for x in range(6)}
print('squares:', squares)

# Invert a dict
original = {'a': 1, 'b': 2, 'c': 3}
inverted = {v: k for k, v in original.items()}
print('inverted:', inverted)

# Filter
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'Dave': 95}
passing = {name: score for name, score in scores.items() if score >= 80}
print('passing:', passing)
```

## 4. defaultdict

```python
from collections import defaultdict

# No KeyError on missing keys
dd = defaultdict(int)
dd['count'] += 1  # starts at 0, no KeyError
dd['count'] += 1
print('count:', dd['count'])

# Word frequency
text = 'the quick brown fox jumps over the lazy dog the'
freq = defaultdict(int)
for word in text.split():
    freq[word] += 1
print('freq:', dict(freq))

# Grouping
groups = defaultdict(list)
data = [('a', 1), ('b', 2), ('a', 3), ('b', 4), ('c', 5)]
for key, val in data:
    groups[key].append(val)
print('groups:', dict(groups))
```

## 5. Counter

```python
from collections import Counter

# Count elements
c = Counter('abracadabra')
print('Counter:', c)
print('most_common(3):', c.most_common(3))

# Arithmetic
c1 = Counter('aab')
c2 = Counter('abc')
print('c1 + c2:', c1 + c2)  # add counts
print('c1 - c2:', c1 - c2)  # subtract (only positive)
print('c1 & c2:', c1 & c2)  # intersection (min)
print('c1 | c2:', c1 | c2)  # union (max)

# Practical: find most common words
text = 'to be or not to be that is the question to be'
word_count = Counter(text.split())
print('Top 3 words:', word_count.most_common(3))
```

## 6. OrderedDict and ChainMap

```python
from collections import OrderedDict, ChainMap

# OrderedDict — useful for move_to_end and order-sensitive equality
od = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
od.move_to_end('a')           # move 'a' to end
print('After move_to_end(a):', list(od.keys()))

od.move_to_end('c', last=False)  # move 'c' to front
print('After move_to_end(c, front):', list(od.keys()))

# Order-sensitive equality
od1 = OrderedDict([('a', 1), ('b', 2)])
od2 = OrderedDict([('b', 2), ('a', 1)])
d1 = {'a': 1, 'b': 2}
d2 = {'b': 2, 'a': 1}
print('OrderedDict equal:', od1 == od2)  # False — order matters
print('dict equal:', d1 == d2)           # True — order ignored
```

```python
# ChainMap — layered lookup without copying
defaults = {'color': 'red', 'size': 'medium', 'debug': False}
user_prefs = {'color': 'blue'}
config = ChainMap(user_prefs, defaults)

print('color:', config['color'])   # 'blue' — user_prefs wins
print('size:', config['size'])     # 'medium' — falls back to defaults

# Writes go to first map
config['theme'] = 'dark'
print('user_prefs after write:', user_prefs)
```

## 7. Merging Dicts

```python
d1 = {'a': 1, 'b': 2}
d2 = {'b': 3, 'c': 4}  # 'b' conflicts

# Python 3.9+ — | operator (rightmost wins)
merged = d1 | d2
print('d1 | d2:', merged)  # {'a': 1, 'b': 3, 'c': 4}

# Python 3.5+ — unpacking
merged2 = {**d1, **d2}
print('{**d1, **d2}:', merged2)

# In-place update
d1_copy = d1.copy()
d1_copy |= d2  # Python 3.9+
print('d1 |= d2:', d1_copy)

# update() method
d1_copy2 = d1.copy()
d1_copy2.update(d2)
print('d1.update(d2):', d1_copy2)
```

## 8. Interview Practice

```python
# Q: Implement word frequency counter
def word_frequency(text):
    freq = {}
    for word in text.lower().split():
        freq[word] = freq.get(word, 0) + 1
    return freq

text = 'to be or not to be that is the question'
print(word_frequency(text))
```

```python
# Q: Two-sum using dict (O(n) solution)
def two_sum(nums, target):
    seen = {}  # value -> index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

print(two_sum([2, 7, 11, 15], 9))   # [0, 1]
print(two_sum([3, 2, 4], 6))        # [1, 2]
```

```python
# Q: Dict ordering — Python 3.7+ guarantees insertion order
d = {}
d['c'] = 3
d['a'] = 1
d['b'] = 2
print('Insertion order preserved:', list(d.keys()))  # ['c', 'a', 'b']

# Q: Why can't you use a list as a dict key?
try:
    d = {[1, 2]: 'value'}
except TypeError as e:
    print('Error:', e)

# Use tuple instead
d = {(1, 2): 'value'}
print('Tuple key works:', d[(1, 2)])
```
