# Python Comprehensions — Interview Preparation

Covers: list/dict/set/generator comprehensions, nested, conditional, performance, and interview Q&A.

## 1. List Comprehensions

```python
# Basic: [expression for item in iterable]
squares = [x**2 for x in range(10)]
print('squares:', squares)

# With filter: [expression for item in iterable if condition]
evens = [x for x in range(20) if x % 2 == 0]
print('evens:', evens)

# With transformation
words = ['  hello  ', ' WORLD ', '  python  ']
cleaned = [w.strip().lower() for w in words]
print('cleaned:', cleaned)
```

## 2. Dict and Set Comprehensions

```python
# Dict comprehension
squares_dict = {x: x**2 for x in range(6)}
print('squares_dict:', squares_dict)

# Invert a dict
original = {'a': 1, 'b': 2, 'c': 3}
inverted = {v: k for k, v in original.items()}
print('inverted:', inverted)

# Set comprehension
unique_squares = {x**2 for x in range(-5, 6)}
print('unique squares:', sorted(unique_squares))
```

## 3. Generator Expressions

```python
import sys

# Generator expression — lazy, memory efficient
gen = (x**2 for x in range(1000))
lst = [x**2 for x in range(1000)]

print(f'Generator size: {sys.getsizeof(gen)} bytes')
print(f'List size: {sys.getsizeof(lst)} bytes')

# Use with sum, max, any, all — no intermediate list
total = sum(x**2 for x in range(1000))
print('sum of squares:', total)

has_large = any(x > 900 for x in range(1000))
print('any > 900:', has_large)

# Generator can only be iterated ONCE
gen = (x for x in range(3))
print('First pass:', list(gen))   # [0, 1, 2]
print('Second pass:', list(gen))  # []  — exhausted!
```

## 4. Nested Comprehensions

```python
# Flatten a matrix
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [x for row in matrix for x in row]
print('Flattened:', flat)
# Read as: for row in matrix: for x in row: yield x

# Transpose a matrix
transposed = [[row[i] for row in matrix] for i in range(3)]
print('Transposed:')
for row in transposed:
    print(' ', row)

# Cartesian product
pairs = [(x, y) for x in [1, 2, 3] for y in ['a', 'b']]
print('Cartesian product:', pairs)
```

## 5. Conditional Comprehensions

```python
# Filter (if at end) — only includes matching elements
evens = [x for x in range(10) if x % 2 == 0]
print('Filter:', evens)  # [0, 2, 4, 6, 8]

# Ternary (if-else in expression) — includes ALL elements, transforms
labels = ['even' if x % 2 == 0 else 'odd' for x in range(6)]
print('Ternary:', labels)

# Both filter AND ternary
result = [x**2 if x > 0 else 0 for x in range(-3, 4) if x != 0]
print('Filter + ternary:', result)
```

## 6. Performance Comparison

```python
import timeit

# For loop vs comprehension vs map
def with_loop():
    result = []
    for x in range(1000):
        result.append(x**2)
    return result

def with_comprehension():
    return [x**2 for x in range(1000)]

def with_map():
    return list(map(lambda x: x**2, range(1000)))

import math
def with_map_named():
    return list(map(math.sqrt, range(1000)))

n = 5000
t_loop = timeit.timeit(with_loop, number=n)
t_comp = timeit.timeit(with_comprehension, number=n)
t_map = timeit.timeit(with_map, number=n)

print(f'for loop:      {t_loop:.4f}s')
print(f'comprehension: {t_comp:.4f}s  ({t_loop/t_comp:.1f}x faster than loop)')
print(f'map+lambda:    {t_map:.4f}s')
```

## 7. Interview Practice

```python
# Q: Variable scoping in comprehensions (Python 3)
x = 'outer'
result = [x for x in range(5)]  # x is scoped to comprehension
print('x after comprehension:', x)  # 'outer' — NOT leaked in Python 3
```

```python
# Q: Walrus operator in comprehension (Python 3.8+)
import math

data = [1, 4, 9, 16, 25, 36]

# Without walrus — sqrt computed twice
result1 = [math.sqrt(x) for x in data if math.sqrt(x) > 3]

# With walrus — sqrt computed once
result2 = [y for x in data if (y := math.sqrt(x)) > 3]

print('result1:', result1)
print('result2:', result2)
```

```python
# Q: Avoid side effects in comprehensions
# BAD — using comprehension just for side effects
# [print(x) for x in range(5)]  # works but bad style

# GOOD — use for loop for side effects
for x in range(5):
    print(x, end=' ')
print()

# Q: Comprehension vs generator for sum
import sys

# Generator — no intermediate list
gen_sum = sum(x**2 for x in range(10**6))

# List comprehension — creates full list in memory
# lst_sum = sum([x**2 for x in range(10**6)])  # uses ~8MB

print('Generator sum:', gen_sum)
```
