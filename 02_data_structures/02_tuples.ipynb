# Python Tuples — Interview Preparation

Covers: creation, unpacking, named tuples, immutability, tuple as dict key, and interview Q&A.

## 1. Tuple Creation

```python
# Various ways to create tuples
empty = ()
single = (42,)      # trailing comma is REQUIRED for single element!
not_tuple = (42)    # this is just an int!
coords = (10, 20)
rgb = (255, 128, 0)
from_list = tuple([1, 2, 3])
from_string = tuple('abc')

print('single:', single, type(single))
print('not_tuple:', not_tuple, type(not_tuple))  # int!
print('from_string:', from_string)

# Packing — parentheses optional
point = 3, 4
print('point:', point, type(point))  # (3, 4) tuple
```

## 2. Tuple Unpacking

```python
# Basic unpacking
x, y = (10, 20)
print(f'x={x}, y={y}')

# Swap without temp variable
a, b = 1, 2
a, b = b, a
print(f'After swap: a={a}, b={b}')

# Extended unpacking with *
first, *rest = (1, 2, 3, 4, 5)
print(f'first={first}, rest={rest}')

*init, last = (1, 2, 3, 4, 5)
print(f'init={init}, last={last}')

first, *middle, last = (1, 2, 3, 4, 5)
print(f'first={first}, middle={middle}, last={last}')
```

```python
# Unpacking in loops
points = [(1, 2), (3, 4), (5, 6)]
for x, y in points:
    print(f'  x={x}, y={y}')

# Ignoring values with _
_, important, _ = (1, 42, 3)
print('important:', important)

# Nested unpacking
(a, b), c = (1, 2), 3
print(f'a={a}, b={b}, c={c}')
```

## 3. Named Tuples

```python
from collections import namedtuple

# Define a named tuple
Point = namedtuple('Point', ['x', 'y'])
Person = namedtuple('Person', 'name age email')  # space-separated string

p = Point(3, 4)
alice = Person('Alice', 30, 'alice@example.com')

# Access by name or index
print('p.x:', p.x, '  p[0]:', p[0])
print('alice.name:', alice.name)
print('alice[1]:', alice[1])

# _replace — returns new namedtuple with changed fields
p2 = p._replace(x=10)
print('p2:', p2)

# _asdict — convert to dict
print('p._asdict():', p._asdict())

# _fields — field names
print('Point._fields:', Point._fields)
```

```python
# Modern: typing.NamedTuple with type annotations and defaults
from typing import NamedTuple

class Point3D(NamedTuple):
    x: float
    y: float
    z: float = 0.0  # default value
    
    def distance_to_origin(self):
        return (self.x**2 + self.y**2 + self.z**2) ** 0.5

p = Point3D(1.0, 2.0)
print('p:', p)
print('z default:', p.z)
print('distance:', p.distance_to_origin())
```

## 4. Immutability and Its Benefits

```python
import sys

# Memory comparison
lst = [1, 2, 3, 4, 5]
tup = (1, 2, 3, 4, 5)
print(f'list size: {sys.getsizeof(lst)} bytes')
print(f'tuple size: {sys.getsizeof(tup)} bytes')

# Speed comparison
import timeit
t_list = timeit.timeit('[1,2,3,4,5]', number=5_000_000)
t_tuple = timeit.timeit('(1,2,3,4,5)', number=5_000_000)
print(f'list creation: {t_list:.3f}s')
print(f'tuple creation: {t_tuple:.3f}s')
print(f'Tuple is ~{t_list/t_tuple:.1f}x faster to create')
```

```python
# Immutability is SHALLOW — mutable elements can still be modified!
t = ([1, 2], [3, 4])
print('Before:', t)

# t[0] = [9, 9]  # TypeError — can't reassign tuple element
t[0].append(99)   # OK — modifying the list object itself
print('After t[0].append(99):', t)

# Such a tuple is NOT hashable
try:
    hash(t)
except TypeError as e:
    print('hash error:', e)
```

## 6. Interview Practice

```python
# Q: What is the difference between (1) and (1,)?
a = (1)
b = (1,)
print(f'(1) type: {type(a)}, value: {a}')   # int
print(f'(1,) type: {type(b)}, value: {b}')  # tuple
```

```python
# Q: Return multiple values from a function
def min_max(lst):
    return min(lst), max(lst)  # returns a tuple

result = min_max([3, 1, 4, 1, 5, 9])
print('result:', result, type(result))

lo, hi = min_max([3, 1, 4, 1, 5, 9])  # unpack
print(f'min={lo}, max={hi}')
```

```python
# Q: When to use tuple vs list?
# Tuple: heterogeneous, fixed, semantic meaning
employee = ('Alice', 30, 'Engineering', 90000)  # record
rgb_red = (255, 0, 0)                           # fixed structure

# List: homogeneous, may change
scores = [95, 87, 92, 78]  # may grow/shrink
names = ['Alice', 'Bob', 'Charlie']  # may be modified

# Tuple in set
visited = set()
visited.add((0, 0))
visited.add((1, 2))
print('visited:', visited)
print('(0,0) in visited:', (0, 0) in visited)
```

```python
# Q: Convert between list and tuple
lst = [3, 1, 4, 1, 5]
tup = tuple(lst)
print('tuple:', tup)

# Sort a tuple (returns a list)
sorted_list = sorted(tup)
sorted_tuple = tuple(sorted(tup))
print('sorted list:', sorted_list)
print('sorted tuple:', sorted_tuple)
```
