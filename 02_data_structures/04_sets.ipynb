# Python Sets — Interview Preparation

Covers: creation, operations, frozenset, set comprehension, time complexity, and interview Q&A.

## 1. Set Creation and Basic Operations

```python
# IMPORTANT: {} creates an empty DICT, not a set!
empty_dict = {}
empty_set = set()
print(type(empty_dict))  # <class 'dict'>
print(type(empty_set))   # <class 'set'>

# Creating sets
s = {1, 2, 3, 4, 5}
s_dups = {1, 2, 2, 3, 3, 3}  # duplicates removed
from_list = set([1, 2, 3, 2, 1])
from_string = set('hello')    # unique characters

print('s_dups:', s_dups)       # {1, 2, 3}
print('from_list:', from_list) # {1, 2, 3}
print('from_string:', from_string)  # {'h', 'e', 'l', 'o'}
```

```python
# Basic operations
s = {1, 2, 3}

s.add(4)       # add element
s.add(2)       # no duplicate, no error
print('After add(4) and add(2):', s)

s.remove(2)    # raises KeyError if not found
s.discard(99)  # no error if not found
print('After remove(2) and discard(99):', s)

print('3 in s:', 3 in s)    # O(1) membership test
print('99 in s:', 99 in s)
```

## 2. Set Operations

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

# Union — all elements from both
print('a | b:', a | b)              # {1, 2, 3, 4, 5, 6}
print('a.union(b):', a.union(b))

# Intersection — elements in both
print('a & b:', a & b)              # {3, 4}
print('a.intersection(b):', a.intersection(b))

# Difference — in a but not b
print('a - b:', a - b)              # {1, 2}
print('b - a:', b - a)              # {5, 6}  — order matters!

# Symmetric difference — in either but not both
print('a ^ b:', a ^ b)              # {1, 2, 5, 6}
print('a.symmetric_difference(b):', a.symmetric_difference(b))
```

```python
# Subset and superset
a = {1, 2}
b = {1, 2, 3, 4}

print('a <= b (subset):', a <= b)         # True
print('a.issubset(b):', a.issubset(b))    # True
print('b >= a (superset):', b >= a)       # True
print('a < b (proper subset):', a < b)   # True (a != b)
print('a < a:', a < a)                   # False — not proper subset

# Disjoint — no common elements
print('{1,2}.isdisjoint({3,4}):', {1,2}.isdisjoint({3,4}))  # True
print('{1,2}.isdisjoint({2,3}):', {1,2}.isdisjoint({2,3}))  # False
```

## 3. Set Comprehension

```python
# Basic set comprehension
squares = {x**2 for x in range(-5, 6)}
print('Unique squares:', sorted(squares))

# Unique characters (case-insensitive)
unique_chars = {c.lower() for c in 'Hello World' if c.isalpha()}
print('Unique chars:', sorted(unique_chars))

# Unique domains from emails
emails = ['alice@gmail.com', 'bob@yahoo.com', 'charlie@gmail.com', 'dave@outlook.com']
domains = {email.split('@')[1] for email in emails}
print('Unique domains:', domains)
```

## 4. frozenset

```python
# frozenset — immutable, hashable set
fs = frozenset([1, 2, 3])
print('frozenset:', fs)

# Can be used as dict key or set member
d = {frozenset([1, 2]): 'pair', frozenset([3, 4]): 'other pair'}
print('dict with frozenset keys:', d)

# Set of frozensets
s = {frozenset([1, 2]), frozenset([3, 4]), frozenset([1, 2])}  # duplicate removed
print('set of frozensets:', s)

# Supports all non-mutating operations
fs1 = frozenset([1, 2, 3])
fs2 = frozenset([3, 4, 5])
print('fs1 | fs2:', fs1 | fs2)
print('fs1 & fs2:', fs1 & fs2)

# Cannot mutate
try:
    fs.add(4)
except AttributeError as e:
    print('Error:', e)
```

## 5. Performance: Set vs List for Membership Testing

```python
import timeit

# O(n) for list vs O(1) for set
big_list = list(range(100_000))
big_set = set(range(100_000))

t_list = timeit.timeit('99_999 in big_list', globals=globals(), number=1000)
t_set = timeit.timeit('99_999 in big_set', globals=globals(), number=1000)

print(f'list "in": {t_list:.4f}s')
print(f'set  "in": {t_set:.6f}s')
print(f'Set is ~{t_list/t_set:.0f}x faster')
```

## 6. Interview Practice

```python
# Q: Remove duplicates from a list
lst = [3, 1, 2, 1, 3, 4]

# Fast but order NOT preserved
unique_unordered = list(set(lst))
print('Unordered unique:', unique_unordered)

# Order preserved (Python 3.7+)
unique_ordered = list(dict.fromkeys(lst))
print('Ordered unique:', unique_ordered)
```

```python
# Q: Find elements unique to one list (symmetric difference)
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

common = set(a) & set(b)
only_in_a = set(a) - set(b)
only_in_b = set(b) - set(a)
unique_to_one = set(a) ^ set(b)

print('Common:', common)
print('Only in a:', only_in_a)
print('Only in b:', only_in_b)
print('Unique to one:', unique_to_one)
```

```python
# Q: Two-sum using set (check existence only)
def two_sum_exists(nums, target):
    seen = set()
    for num in nums:
        if target - num in seen:  # O(1) lookup
            return True
        seen.add(num)
    return False

print(two_sum_exists([2, 7, 11, 15], 9))   # True
print(two_sum_exists([1, 2, 3], 10))       # False

# Q: remove() vs discard()
s = {1, 2, 3}
s.discard(99)  # no error
try:
    s.remove(99)  # KeyError!
except KeyError as e:
    print('remove(99) KeyError:', e)
```
