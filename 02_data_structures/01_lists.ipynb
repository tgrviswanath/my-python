## 1. List Creation

## 2. Indexing and Slicing

```python
lst = [10, 20, 30, 40, 50]

# Positive and negative indexing
print('lst[0]:', lst[0])    # 10
print('lst[-1]:', lst[-1])  # 50
print('lst[-2]:', lst[-2])  # 40

# Slicing: [start:stop:step] — stop is exclusive
print('lst[1:4]:', lst[1:4])   # [20, 30, 40]
print('lst[:3]:', lst[:3])     # [10, 20, 30]
print('lst[2:]:', lst[2:])     # [30, 40, 50]
print('lst[::2]:', lst[::2])   # [10, 30, 50] — every other
print('lst[::-1]:', lst[::-1]) # [50, 40, 30, 20, 10] — reversed
```

```python
# Slice assignment
lst = [10, 20, 30, 40, 50]
lst[1:3] = [200, 300]  # replace elements at index 1 and 2
print('After slice assignment:', lst)  # [10, 200, 300, 40, 50]

lst[::2] = [0, 0, 0]   # replace every other element
print('After step assignment:', lst)
```

## 3. Core Methods

```python
# append vs extend — key difference!
lst = [1, 2, 3]

lst.append([4, 5])   # adds [4,5] as a SINGLE element
print('After append([4,5]):', lst)  # [1, 2, 3, [4, 5]]

lst = [1, 2, 3]
lst.extend([4, 5])   # adds each element individually
print('After extend([4,5]):', lst)  # [1, 2, 3, 4, 5]
```

```python
# insert, remove, pop
lst = [1, 2, 3, 4]

lst.insert(1, 99)   # insert 99 at index 1 — O(n)
print('After insert(1, 99):', lst)

lst.remove(99)      # remove first occurrence of 99 — O(n)
print('After remove(99):', lst)

val = lst.pop()     # remove and return last element — O(1)
print(f'pop() returned {val}, lst is now:', lst)

val = lst.pop(0)    # remove and return element at index 0 — O(n)
print(f'pop(0) returned {val}, lst is now:', lst)
```

```python
# sort, reverse, index, count
lst = [3, 1, 4, 1, 5, 9, 2, 6]

print('count(1):', lst.count(1))   # 2
print('index(5):', lst.index(5))   # 4

lst.sort()                          # in-place sort — O(n log n)
print('After sort():', lst)

lst.reverse()                       # in-place reverse — O(n)
print('After reverse():', lst)

lst.sort(key=lambda x: -x)          # sort descending via key
print('Sorted descending:', lst)
```

## 4. List as Stack and Queue

```python
# Stack (LIFO) — use append/pop — both O(1)
stack = []
stack.append('a')  # push
stack.append('b')
stack.append('c')
print('Stack:', stack)
print('Pop:', stack.pop())  # 'c'
print('Pop:', stack.pop())  # 'b'
print('Stack after pops:', stack)
```

```python
# Queue (FIFO) — use collections.deque, NOT list!
# list.pop(0) is O(n) — avoid for queues
from collections import deque

queue = deque()
queue.append('first')   # enqueue — O(1)
queue.append('second')
queue.append('third')
print('Queue:', queue)
print('Dequeue:', queue.popleft())  # 'first' — O(1)
print('Dequeue:', queue.popleft())  # 'second'
print('Queue after dequeues:', queue)
```

## 5. sort() vs sorted()

```python
# sort() — in-place, returns None
lst = [3, 1, 4, 1, 5]
result = lst.sort()
print('lst.sort() returns:', result)  # None!
print('lst after sort():', lst)       # modified in place

# sorted() — returns new list, original unchanged
lst = [3, 1, 4, 1, 5]
new_lst = sorted(lst)
print('sorted() returns:', new_lst)
print('original unchanged:', lst)

# Common mistake!
lst = [3, 1, 2]
lst = lst.sort()  # lst is now None!
print('After lst = lst.sort():', lst)  # None — BUG!
```

```python
# key parameter and stability
words = ['banana', 'apple', 'cherry', 'date']
print('By length:', sorted(words, key=len))
print('By last char:', sorted(words, key=lambda s: s[-1]))

# Stability: equal elements keep original order
data = [('Alice', 2), ('Bob', 1), ('Alice', 1)]
result = sorted(data, key=lambda x: x[0])
print('Stable sort by name:', result)
# Both Alices: ('Alice', 2) comes before ('Alice', 1) — original order preserved
```

## 6. Shallow vs Deep Copy

```python
import copy

# Shallow copy — nested objects are SHARED
original = [[1, 2], [3, 4]]
shallow = original.copy()  # or original[:] or list(original)

shallow[0].append(99)  # modifies the shared inner list!
print('original after shallow copy mutation:', original)  # [[1, 2, 99], [3, 4]]
print('shallow:', shallow)                                # [[1, 2, 99], [3, 4]]

# Deep copy — fully independent
original = [[1, 2], [3, 4]]
deep = copy.deepcopy(original)

deep[0].append(99)
print('original after deep copy mutation:', original)  # [[1, 2], [3, 4]] — unchanged!
print('deep:', deep)                                   # [[1, 2, 99], [3, 4]]
```

## 7. Time Complexity Demo

```python
import timeit

# append is O(1) amortized — fast
t_append = timeit.timeit(
    'lst.append(1)',
    setup='lst = list(range(10000))',
    number=100000
)

# insert(0, x) is O(n) — slow for large lists
t_insert = timeit.timeit(
    'lst.insert(0, 1)',
    setup='lst = list(range(10000))',
    number=100000
)

print(f'append: {t_append:.4f}s')
print(f'insert(0): {t_insert:.4f}s')
print(f'insert is ~{t_insert/t_append:.0f}x slower than append')
```

```python
# 'in' operator: O(n) for list, O(1) for set
big_list = list(range(100000))
big_set = set(range(100000))

t_list = timeit.timeit('99999 in big_list', globals=globals(), number=1000)
t_set = timeit.timeit('99999 in big_set', globals=globals(), number=1000)

print(f'list "in": {t_list:.4f}s')
print(f'set "in":  {t_set:.6f}s')
print(f'Set is ~{t_list/t_set:.0f}x faster for membership testing')
```

## 8. List Comprehensions

```python
# Basic comprehension
squares = [x**2 for x in range(10)]
print('Squares:', squares)

# With filter
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print('Even squares:', even_squares)

# Flatten matrix
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [x for row in matrix for x in row]
print('Flattened:', flat)

# Comprehension vs map — performance comparison
t_comp = timeit.timeit('[x**2 for x in range(1000)]', number=10000)
t_map = timeit.timeit('list(map(lambda x: x**2, range(1000)))', number=10000)
print(f'Comprehension: {t_comp:.4f}s')
print(f'map+lambda:    {t_map:.4f}s')
```

## 9. Interview Practice

```python
# Q: Remove duplicates preserving order
lst = [3, 1, 2, 1, 3, 4, 2]

# Method 1: dict.fromkeys (Python 3.7+ preserves order)
unique = list(dict.fromkeys(lst))
print('Unique (ordered):', unique)  # [3, 1, 2, 4]

# Method 2: seen set
seen = set()
unique2 = [x for x in lst if not (x in seen or seen.add(x))]
print('Unique (seen set):', unique2)
```

```python
# Q: What is the difference between del, remove(), and pop()?
lst = [10, 20, 30, 40, 50]

# del — by index or slice, no return value
del lst[1]
print('After del lst[1]:', lst)  # [10, 30, 40, 50]

# remove — by VALUE, removes first occurrence
lst.remove(30)
print('After remove(30):', lst)  # [10, 40, 50]

# pop — by index, returns the removed element
val = lst.pop(0)
print(f'pop(0) returned {val}, lst:', lst)  # 10, [40, 50]
```

```python
# Q: Pitfall — a = b = [] vs a, b = [], []
a = b = []          # SAME object!
a.append(1)
print('b after a.append(1):', b)  # [1] — b is affected!

a, b = [], []       # DIFFERENT objects
a.append(1)
print('b after a.append(1):', b)  # [] — b is unaffected
```

```python
# Q: Implement a stack and verify LIFO behavior
class Stack:
    def __init__(self):
        self._data = []
    
    def push(self, item):
        self._data.append(item)  # O(1)
    
    def pop(self):
        if not self._data:
            raise IndexError('Stack is empty')
        return self._data.pop()  # O(1)
    
    def peek(self):
        return self._data[-1] if self._data else None
    
    def __len__(self):
        return len(self._data)
    
    def __repr__(self):
        return f'Stack({self._data})'

s = Stack()
for i in [1, 2, 3]:
    s.push(i)
print('Stack:', s)
print('Pop order (LIFO):', [s.pop() for _ in range(len(s))])
```
