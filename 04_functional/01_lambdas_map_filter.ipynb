# Python Lambdas, map, filter, reduce — Interview Preparation

Covers: lambda syntax, map/filter/reduce, sorted with key, when to use lambda vs def, and interview Q&A.

## 1. Lambda Syntax

```python
# Lambda: anonymous, single-expression function
square = lambda x: x ** 2
add = lambda x, y: x + y
clamp = lambda x, lo, hi: max(lo, min(hi, x))

print('square(5):', square(5))
print('add(3, 4):', add(3, 4))
print('clamp(15, 0, 10):', clamp(15, 0, 10))

# Ternary in lambda
abs_val = lambda x: x if x >= 0 else -x
print('abs_val(-7):', abs_val(-7))

# Immediately invoked
result = (lambda x, y: x + y)(3, 4)
print('Immediately invoked:', result)
```

```python
# Lambda vs def — PEP 8 says don't assign lambda to a variable
# BAD (PEP 8 violation):
f = lambda x: x ** 2

# GOOD:
def f(x):
    return x ** 2

# Lambda is fine as an argument:
data = [3, 1, 4, 1, 5, 9]
sorted_data = sorted(data, key=lambda x: -x)  # OK
print('Sorted descending:', sorted_data)
```

## 2. map()

```python
nums = [1, 2, 3, 4, 5]

# map with lambda
squares = list(map(lambda x: x**2, nums))
print('squares:', squares)

# map with named function (faster — no lambda overhead)
import math
roots = list(map(math.sqrt, nums))
print('roots:', roots)

# map with multiple iterables
a = [1, 2, 3]
b = [10, 20, 30]
sums = list(map(lambda x, y: x + y, a, b))
print('sums:', sums)

# map is LAZY — returns iterator
gen = map(lambda x: x**2, range(10**6))
print('type:', type(gen))  # <class 'map'>
print('first:', next(gen)) # 0 — computed on demand
```

```python
# map vs list comprehension — performance
import timeit

t_comp = timeit.timeit('[x**2 for x in range(1000)]', number=10000)
t_map_lambda = timeit.timeit('list(map(lambda x: x**2, range(1000)))', number=10000)
t_map_named = timeit.timeit('list(map(abs, range(1000)))', number=10000)

print(f'comprehension:  {t_comp:.4f}s')
print(f'map + lambda:   {t_map_lambda:.4f}s')
print(f'map + named fn: {t_map_named:.4f}s  (fastest!)')
```

## 3. filter()

```python
nums = [1, -2, 3, -4, 5, -6, 0]

# filter with lambda
positives = list(filter(lambda x: x > 0, nums))
print('positives:', positives)

# filter with named function
evens = list(filter(lambda x: x % 2 == 0, nums))
print('evens:', evens)

# filter(None, iterable) — removes falsy values
mixed = [0, 1, '', 'hello', None, [], [1, 2], False, True]
truthy = list(filter(None, mixed))
print('truthy:', truthy)

# Equivalent comprehension
truthy2 = [x for x in mixed if x]
print('truthy2:', truthy2)
```

## 4. reduce()

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

# Sum: 1+2=3, 3+3=6, 6+4=10, 10+5=15
total = reduce(lambda acc, x: acc + x, nums)
print('sum:', total)

# Product
product = reduce(lambda acc, x: acc * x, nums)
print('product:', product)

# Max
maximum = reduce(lambda a, b: a if a > b else b, nums)
print('max:', maximum)

# With initializer
total_with_init = reduce(lambda acc, x: acc + x, nums, 100)
print('sum with init=100:', total_with_init)

# Flatten
nested = [[1, 2], [3, 4], [5, 6]]
flat = reduce(lambda acc, x: acc + x, nested, [])
print('flattened:', flat)

# Note: prefer sum(), max(), min() for common reductions
print('sum():', sum(nums))
print('max():', max(nums))
```

## 5. sorted() with key

```python
from operator import itemgetter, attrgetter

# Sort by length
words = ['banana', 'apple', 'cherry', 'date']
print('By length:', sorted(words, key=len))

# Sort tuples by second element
people = [('Alice', 30), ('Bob', 25), ('Charlie', 35)]
print('By age:', sorted(people, key=lambda p: p[1]))
print('By age (itemgetter):', sorted(people, key=itemgetter(1)))  # faster

# Multiple sort criteria — tuple key
students = [('Alice', 'B', 3.8), ('Bob', 'A', 3.5), ('Charlie', 'A', 3.9)]
# Sort by grade ascending, then GPA descending
result = sorted(students, key=lambda s: (s[1], -s[2]))
print('By grade then GPA desc:', result)

# Case-insensitive sort
mixed_case = ['Banana', 'apple', 'Cherry', 'date']
print('Case-insensitive:', sorted(mixed_case, key=str.lower))
```

## 6. Interview Practice

```python
# Q: Implement a pipeline using map/filter
data = ['  Hello  ', ' WORLD ', '  python  ', '  hi  ', '  a  ']

# Pipeline: strip → filter long words → uppercase
result = list(
    map(str.upper,
        filter(lambda s: len(s) > 3,
               map(str.strip, data)))
)
print('Pipeline result:', result)

# Equivalent comprehension (cleaner)
result2 = [s.strip().upper() for s in data if len(s.strip()) > 3]
print('Comprehension result:', result2)
```

```python
# Q: Partial application with lambda
def make_multiplier(n):
    return lambda x: x * n

double = make_multiplier(2)
triple = make_multiplier(3)

print('double(5):', double(5))
print('triple(5):', triple(5))
print('list(map(double, range(5))):', list(map(double, range(5))))

# Q: Sort objects by multiple attributes
records = [
    {'name': 'Alice', 'dept': 'Eng', 'salary': 90000},
    {'name': 'Bob', 'dept': 'HR', 'salary': 70000},
    {'name': 'Charlie', 'dept': 'Eng', 'salary': 85000},
    {'name': 'Dave', 'dept': 'HR', 'salary': 75000},
]

# Sort by dept ascending, then salary descending
sorted_records = sorted(records, key=lambda r: (r['dept'], -r['salary']))
for r in sorted_records:
    print(f"  {r['dept']}: {r['name']} ${r['salary']}")
```
