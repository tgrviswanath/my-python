## 1. Basic Types

```python
# Integer — arbitrary precision
x = 42
big = 2 ** 100
print(type(x), x)
print(type(big), big)
```

```python
# Float
pi = 3.14159
sci = 1.5e-3
print(type(pi), pi)
print(type(sci), sci)
print(0.1 + 0.2)          # floating point imprecision
print(round(0.1 + 0.2, 2)) # fix with round()
```

```python
# Bool — subclass of int
print(type(True), isinstance(True, int))
print(True + True)   # 2
print(True * 5)      # 5
print(False + 1)     # 1
```

```python
# NoneType
nothing = None
print(type(nothing))
print(nothing is None)   # always use 'is' for None check
print(nothing == None)   # works but not idiomatic
```

## 2. Variable Assignment

```python
# Multiple assignment
a = b = c = 0
print(a, b, c)

# Tuple unpacking
x, y = 10, 20
print(x, y)

# Swap without temp variable
x, y = y, x
print(x, y)

# Starred unpacking
first, *middle, last = [1, 2, 3, 4, 5]
print(first, middle, last)
```

## 3. Type Checking & Conversion

```python
# type() vs isinstance()
print(type(42) == int)          # True
print(isinstance(42, int))      # True
print(isinstance(True, int))    # True — bool is subclass of int
print(type(True) == int)        # False — type() is exact
```

```python
# Type conversion
print(int('42'))       # 42
print(int(3.9))        # 3 — truncates, does NOT round
print(float('3.14'))   # 3.14
print(str(100))        # '100'
print(bool(0))         # False
print(bool(''))        # False
print(bool([]))        # False
print(bool('hello'))   # True
```

## 4. Identity vs Equality

```python
# == compares value, 'is' compares identity (id)
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)   # True  — same value
print(a is b)   # False — different objects
print(a is c)   # True  — same object
print(id(a), id(b), id(c))
```

```python
# CPython small int caching (-5 to 256)
x = 256
y = 256
print(x is y)   # True — cached

x = 257
y = 257
print(x is y)   # False — not cached (implementation detail)
```

## 5. Division Types

```python
print(7 / 2)    # 3.5  — true division, always float
print(7 // 2)   # 3    — floor division
print(-7 // 2)  # -4   — floor (toward -inf), NOT truncate
print(7 % 2)    # 1    — modulo
print(-7 % 2)   # 1    — result has sign of divisor
print(divmod(7, 2))  # (3, 1) — quotient and remainder
```

## 6. LEGB Scope Rule

```python
x = 'global'

def outer():
    x = 'enclosing'
    def inner():
        x = 'local'
        print('inner:', x)
    inner()
    print('outer:', x)

outer()
print('module:', x)
```

```python
# global keyword
count = 0

def increment():
    global count
    count += 1

increment()
increment()
print(count)  # 2
```

```python
# nonlocal keyword
def make_counter():
    n = 0
    def inc():
        nonlocal n
        n += 1
        return n
    return inc

counter = make_counter()
print(counter())  # 1
print(counter())  # 2
print(counter())  # 3
```

## 7. Mutable vs Immutable

```python
# Immutable — reassignment creates new object
s = 'hello'
print(id(s))
s = s + ' world'  # new string object
print(id(s))      # different id

# Mutable — in-place modification
lst = [1, 2, 3]
print(id(lst))
lst.append(4)     # same object
print(id(lst))    # same id
```

```python
# Gotcha: mutable default argument
def bad_append(item, lst=[]):   # DON'T do this
    lst.append(item)
    return lst

print(bad_append(1))  # [1]
print(bad_append(2))  # [1, 2] — shared default!

def good_append(item, lst=None):  # correct pattern
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(good_append(1))  # [1]
print(good_append(2))  # [2] — fresh list each time
```
