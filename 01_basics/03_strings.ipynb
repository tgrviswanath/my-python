# 03 — Strings


## 1. String Creation

```python
s1 = 'single'
s2 = "double"
s3 = '''multi
line'''
s4 = r'raw\nno escape'
s5 = f'computed: {2**10}'
print(s1, s2, s3, s4, s5)
```

## 2. String Methods

```python
s = '  Hello, World!  '
print(s.strip())          # 'Hello, World!'
print(s.upper())          # '  HELLO, WORLD!  '
print(s.lower())          # '  hello, world!  '
print(s.replace('World', 'Python'))
print(s.find('World'))    # 9
print(s.count('l'))       # 3
print(s.startswith('  Hello'))  # True
```

```python
# split and join
csv = 'a,b,c,d'
parts = csv.split(',')
print(parts)              # ['a', 'b', 'c', 'd']
print(' | '.join(parts))  # 'a | b | c | d'

# split with limit
print('a:b:c'.split(':', 1))  # ['a', 'b:c']

# splitlines
print('line1\nline2\nline3'.splitlines())  # ['line1', 'line2', 'line3']
```

## 3. String Formatting

```python
name, score, pi = 'Alice', 95.678, 3.14159

# f-strings
print(f'{name} scored {score:.1f}%')
print(f'{pi:.4f}')           # 3.1416
print(f'{1000000:,}')        # 1,000,000
print(f'{42:08b}')           # 00101010
print(f'{42:#010x}')         # 0x0000002a
print(f'{name:>20}')         # right-align in 20 chars
print(f'{name:^20}')         # center
print(f'{name:-<20}')        # left-align, fill with -
```

```python
# Expressions in f-strings
x = 10
print(f'{x**2 + 1}')         # 101
print(f'{"hello".upper()}')  # HELLO
print(f'{[i**2 for i in range(5)]}')  # [0, 1, 4, 9, 16]

# Debug format (Python 3.8+)
value = 42
print(f'{value=}')  # value=42
```

## 4. Slicing

```python
s = 'Hello, World!'
print(s[0])       # H
print(s[-1])      # !
print(s[0:5])     # Hello
print(s[7:])      # World!
print(s[:5])      # Hello
print(s[::2])     # Hlo ol!
print(s[::-1])    # !dlroW ,olleH  (reverse)
print(s[7:12:2])  # Wrd
```

## 5. String Algorithms (Interview)

```python
# Reverse a string
s = 'hello'
print(s[::-1])                    # olleh
print(''.join(reversed(s)))       # olleh
```

```python
# Palindrome check
def is_palindrome(s):
    s = s.lower()
    s = ''.join(c for c in s if c.isalnum())
    return s == s[::-1]

print(is_palindrome('racecar'))                    # True
print(is_palindrome('A man a plan a canal Panama')) # True
print(is_palindrome('hello'))                      # False
```

```python
# Anagram check
from collections import Counter

def is_anagram(s1, s2):
    return Counter(s1.lower()) == Counter(s2.lower())

print(is_anagram('listen', 'silent'))  # True
print(is_anagram('hello', 'world'))    # False
```

```python
# Count character frequency
from collections import Counter

s = 'hello world'
freq = Counter(s)
print(freq.most_common(3))  # [('l', 3), ('o', 2), (' ', 1)]
```

```python
# String concatenation performance
import time

n = 10000
words = ['word'] * n

# Bad: O(n^2)
start = time.perf_counter()
result = ''
for w in words:
    result += w
bad_time = time.perf_counter() - start

# Good: O(n)
start = time.perf_counter()
result = ''.join(words)
good_time = time.perf_counter() - start

print(f'Concatenation: {bad_time*1000:.2f}ms')
print(f'join():        {good_time*1000:.2f}ms')
```

```python
# Find all substrings
def find_all(s, sub):
    positions = []
    start = 0
    while (idx := s.find(sub, start)) != -1:
        positions.append(idx)
        start = idx + 1
    return positions

print(find_all('abcabcabc', 'abc'))  # [0, 3, 6]
```
