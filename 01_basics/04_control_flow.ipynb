## 1. if / elif / else

```python
def classify(n):
    if n > 0:
        return 'positive'
    elif n < 0:
        return 'negative'
    else:
        return 'zero'

for n in [5, -3, 0]:
    print(f'{n:3} -> {classify(n)}')
```

```python
# Ternary expression
x = 7
parity = 'even' if x % 2 == 0 else 'odd'
print(parity)

# Nested ternary (use sparingly)
score = 85
grade = 'A' if score >= 90 else 'B' if score >= 80 else 'C'
print(grade)
```

## 2. for Loops

```python
# range variants
print(list(range(5)))          # [0, 1, 2, 3, 4]
print(list(range(2, 8)))       # [2, 3, 4, 5, 6, 7]
print(list(range(0, 10, 2)))   # [0, 2, 4, 6, 8]
print(list(range(10, 0, -2)))  # [10, 8, 6, 4, 2]

# range is lazy
r = range(10**9)
print(type(r))       # <class 'range'>
print(999999999 in r) # True — O(1)!
```

```python
# enumerate
fruits = ['apple', 'banana', 'cherry']
for i, fruit in enumerate(fruits, start=1):
    print(f'{i}. {fruit}')
```

```python
# zip
names = ['Alice', 'Bob', 'Charlie']
scores = [95, 87, 92]
grades = ['A', 'B', 'A']

for name, score, grade in zip(names, scores, grades):
    print(f'{name}: {score} ({grade})')

# zip stops at shortest
print(list(zip([1,2,3], ['a','b'])))  # [(1,'a'), (2,'b')]
```

```python
# zip_longest
from itertools import zip_longest
for a, b in zip_longest([1,2,3], ['x','y'], fillvalue='?'):
    print(a, b)
```

## 3. break, continue, else

```python
# break
print('break:')
for i in range(10):
    if i == 5:
        break
    print(i, end=' ')
print()

# continue
print('continue:')
for i in range(10):
    if i % 2 == 0:
        continue
    print(i, end=' ')
print()
```

```python
# for...else — else runs only if no break
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Using for...else pattern
def is_prime_v2(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            break
    else:
        return True  # no break — no divisor found
    return False

primes = [n for n in range(2, 30) if is_prime(n)]
print(primes)
```

## 4. while Loops

```python
# Binary search using while
def binary_search(arr, target):
    lo, hi = 0, len(arr) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
    return -1

arr = list(range(0, 20, 2))  # [0, 2, 4, ..., 18]
print(arr)
print(binary_search(arr, 10))  # 5
print(binary_search(arr, 7))   # -1
```

## 5. match / case (Python 3.10+)

```python
import sys
if sys.version_info >= (3, 10):
    def http_status(status):
        match status:
            case 200:
                return 'OK'
            case 404:
                return 'Not Found'
            case 500 | 503:
                return 'Server Error'
            case code if 400 <= code < 500:
                return f'Client Error {code}'
            case _:
                return 'Unknown'

    for code in [200, 404, 500, 403, 999]:
        print(f'{code}: {http_status(code)}')
else:
    print('match/case requires Python 3.10+')
```

```python
# Structural pattern matching
if sys.version_info >= (3, 10):
    def process_point(point):
        match point:
            case (0, 0):
                return 'Origin'
            case (x, 0):
                return f'X-axis at {x}'
            case (0, y):
                return f'Y-axis at {y}'
            case (x, y):
                return f'Point ({x}, {y})'

    for p in [(0,0), (3,0), (0,4), (1,2)]:
        print(f'{p} -> {process_point(p)}')
```

## 6. Interview Patterns

```python
# FizzBuzz — classic interview question
for i in range(1, 21):
    if i % 15 == 0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

```python
# Pythonic FizzBuzz
result = [
    'FizzBuzz' if i % 15 == 0
    else 'Fizz' if i % 3 == 0
    else 'Buzz' if i % 5 == 0
    else str(i)
    for i in range(1, 21)
]
print(result)
```
