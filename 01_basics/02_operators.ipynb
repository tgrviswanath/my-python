# 02 — Operators


## 1. Arithmetic

```python
print(7 + 2)   # 9
print(7 - 2)   # 5
print(7 * 2)   # 14
print(7 / 2)   # 3.5  — always float
print(7 // 2)  # 3    — floor division
print(-7 // 2) # -4   — floor toward -inf
print(7 % 2)   # 1
print(2 ** 10) # 1024
```

## 2. Comparison & Chaining

```python
x = 5
print(1 < x < 10)       # True — chained comparison
print(0 <= x <= 5)      # True
print(x == 5 == 5.0)    # True — chained
print(1 < x > 3)        # True — both conditions checked
```

## 3. Logical Operators — Short Circuit

```python
# and returns first falsy or last value
print(0 and 'hello')     # 0
print(1 and 'hello')     # 'hello'
print('' and 'hello')    # ''

# or returns first truthy or last value
print(0 or 'default')    # 'default'
print('val' or 'default') # 'val'
print(None or 0 or '')   # ''

# Practical: default value pattern
name = None
display = name or 'Anonymous'
print(display)  # 'Anonymous'
```

## 4. Bitwise Operators

```python
a, b = 0b1010, 0b1100  # 10, 12
print(f'a     = {a:04b} = {a}')
print(f'b     = {b:04b} = {b}')
print(f'a & b = {a&b:04b} = {a&b}')  # AND
print(f'a | b = {a|b:04b} = {a|b}')  # OR
print(f'a ^ b = {a^b:04b} = {a^b}')  # XOR
print(f'~a    = {~a}')                # NOT
print(f'a<<1  = {a<<1:04b} = {a<<1}') # left shift
print(f'a>>1  = {a>>1:04b} = {a>>1}') # right shift
```

```python
# Practical: permission flags
READ  = 0b001  # 1
WRITE = 0b010  # 2
EXEC  = 0b100  # 4

perms = 0
perms |= READ | WRITE   # grant READ and WRITE
print(f'perms: {perms:03b}')  # 011

print('can read:', bool(perms & READ))   # True
print('can exec:', bool(perms & EXEC))   # False

perms &= ~WRITE  # revoke WRITE
print(f'after revoke: {perms:03b}')  # 001

# Check even/odd
for n in range(6):
    print(f'{n} is {"even" if n & 1 == 0 else "odd"}')
```

## 5. Walrus Operator := (Python 3.8+)

```python
import re

# Without walrus
data = [1, 5, 3, 8, 2, 9, 4]
filtered = []
for x in data:
    y = x * 2
    if y > 10:
        filtered.append(y)
print(filtered)

# With walrus — cleaner
filtered2 = [y for x in data if (y := x * 2) > 10]
print(filtered2)
```

```python
# Walrus in while loop
import io
fake_file = io.StringIO('line1\nline2\nline3\n')

while line := fake_file.readline():
    print(repr(line.strip()))
```

## 6. Operator Precedence

```python
# ** is right-associative
print(2 ** 3 ** 2)    # 2**(3**2) = 2**9 = 512
print((2 ** 3) ** 2)  # 8**2 = 64

# Precedence examples
print(2 + 3 * 4)      # 14 (not 20)
print(not 1 + 1 == 2) # False: not (1+1==2) = not True
print(True or False and False)  # True: or has lower precedence than and
```

## 7. Operator Overloading

```python
class Vector:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __rmul__(self, scalar):  # scalar * vector
        return self.__mul__(scalar)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f'Vector({self.x}, {self.y})'

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)   # Vector(4, 6)
print(v1 * 3)    # Vector(3, 6)
print(3 * v1)    # Vector(3, 6) — uses __rmul__
print(v1 == Vector(1, 2))  # True
```
