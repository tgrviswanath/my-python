## 1. String Internals — Memory by Encoding Kind

```python
import sys

print("=== PEP 393 Flexible String Representation ===")
samples = [
    ("empty",         ""),
    ("ASCII hello",   "hello"),
    ("Latin-1 cafe",  "caf\u00e9"),
    ("UCS-2 Chinese", "\u4e2d\u6587"),
    ("UCS-4 Emoji",   "\U0001F600"),
    ("Mixed UCS-4",   "hello\U0001F600"),
]
print(f"{'String':20s} {'len':>5} {'getsizeof':>10}")
print("-" * 38)
for name, s in samples:
    print(f"{name:20s} {len(s):>5} {sys.getsizeof(s):>10}")
print()
s1 = "hello"
s2 = "hello\U0001F600"
print(f"'hello'        : {sys.getsizeof(s1)} bytes")
print(f"'hello' + emoji: {sys.getsizeof(s2)} bytes  ({sys.getsizeof(s2)/sys.getsizeof(s1):.1f}x larger)")
```

## 3. String Formatting Performance Benchmark

```python
import timeit

name = "Alice"
score = 95.678
N = 1_000_000

t_fstr   = timeit.timeit(lambda: f"Hello, {name}! Score: {score:.2f}", number=N)
t_format = timeit.timeit(lambda: "Hello, {}! Score: {:.2f}".format(name, score), number=N)
t_pct    = timeit.timeit(lambda: "Hello, %s! Score: %.2f" % (name, score), number=N)

print(f"Formatting {N:,} strings:")
print(f"  f-string:  {t_fstr:.3f}s  (1.00x baseline)")
print(f"  .format(): {t_format:.3f}s  ({t_format/t_fstr:.2f}x slower)")
print(f"  % format:  {t_pct:.3f}s  ({t_pct/t_fstr:.2f}x slower)")
print()
pi = 3.14159265358979
n = 1_234_567
print("=== f-string format spec examples ===")
print(f"  pi:.4f   = {pi:.4f}")
print(f"  n:,      = {n:,}")
print(f"  0.75:.1% = {0.75:.1%}")
print(f"  42:08b   = {42:08b}")
print(f"  42:#010x = {42:#010x}")
```

## 4. Regular Expressions

```python
import re

text = "Contact: alice@example.com, bob@company.org. Phone: 555-123-4567. Date: 2024-01-15."

print("=== re.findall ===")
emails = re.findall(r'\b[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}\b', text)
print(f"Emails: {emails}")

phones = re.findall(r'\b\d{3}[-.]?\d{3}[-.]?\d{4}\b', text)
print(f"Phones: {phones}")

dates = re.findall(r'\b\d{4}-\d{2}-\d{2}\b', text)
print(f"Dates:  {dates}")

print()
print("=== Named Groups ===")
m = re.search(r'(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})', text)
if m:
    print(f"Full match: {m.group(0)}")
    print(f"Named:      {m.groupdict()}")

print()
print("=== re.sub ===")
redacted = re.sub(r'\b[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}\b', '[EMAIL]', text)
print(f"Redacted: {redacted}")

print()
print("=== re.split ===")
parts = re.split(r'[,;.\s]+', 'one, two;  three.four five')
parts = [p for p in parts if p]
print(f"Split: {parts}")
```

```python
import re, timeit

print("=== match vs search vs fullmatch ===")
pattern = r'\d+'
tests = ["123", "abc123", "123abc", "abc"]
print(f"{'String':10s} {'match':8s} {'search':8s} {'fullmatch':10s}")
print("-" * 40)
for s in tests:
    m  = bool(re.match(pattern, s))
    sr = bool(re.search(pattern, s))
    fm = bool(re.fullmatch(pattern, s))
    print(f"{s!r:10s} {str(m):8s} {str(sr):8s} {str(fm):10s}")

print()
print("=== Compiled vs Uncompiled ===")
text2 = "Contact alice@example.com for info " * 100
pat_str = r'\b[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}\b'
pat_cmp = re.compile(pat_str)
N = 5000
t1 = timeit.timeit(lambda: re.findall(pat_str, text2), number=N)
t2 = timeit.timeit(lambda: pat_cmp.findall(text2), number=N)
print(f"Uncompiled ({N} calls): {t1:.3f}s")
print(f"Compiled   ({N} calls): {t2:.3f}s")
print(f"Speedup: {t1/t2:.2f}x")
```

## 5. Encoding and Decoding

```python
print("=== Encoding str to bytes ===")
text = "caf\u00e9"
print(f"Original: {text!r}")
print()
for enc in ["utf-8", "utf-16", "latin-1"]:
    try:
        encoded = text.encode(enc)
        print(f"{enc:10s}: {len(encoded):3d} bytes -> {encoded}")
    except UnicodeEncodeError as e:
        print(f"{enc:10s}: ERROR - {e}")

print()
print("=== Error Handling Modes ===")
for mode in ["strict", "ignore", "replace", "xmlcharrefreplace", "backslashreplace"]:
    try:
        result = text.encode("ascii", errors=mode)
        print(f"{mode:20s}: {result}")
    except UnicodeEncodeError:
        print(f"{mode:20s}: UnicodeEncodeError")

print()
import base64
original = b"Hello, World!"
encoded = base64.b64encode(original)
decoded = base64.b64decode(encoded)
print(f"Original: {original}")
print(f"Base64:   {encoded}")
print(f"Decoded:  {decoded}")
print(f"Match:    {original == decoded}")
```

## 6. String Building Performance

```python
import timeit, io

words = ["word"] * 1000
N = 500

def concat_plus():
    result = ""
    for w in words:
        result += w + " "
    return result

def join_method():
    return " ".join(words)

def stringio_method():
    buf = io.StringIO()
    for w in words:
        buf.write(w)
        buf.write(" ")
    return buf.getvalue()

t_plus = timeit.timeit(concat_plus, number=N)
t_join = timeit.timeit(join_method, number=N)
t_sio  = timeit.timeit(stringio_method, number=N)

print(f"String building (1000 words, {N} iterations):")
print(f"  += concatenation: {t_plus:.3f}s  (O(n^2))")
print(f"  join():           {t_join:.3f}s  (O(n)) -- {t_plus/t_join:.1f}x faster")
print(f"  io.StringIO:      {t_sio:.3f}s  (O(n)) -- {t_plus/t_sio:.1f}x faster")
```

## 7. String Algorithms

```python
from collections import Counter

def reverse_string(s):
    return s[::-1]

def is_palindrome(s):
    cleaned = "".join(c.lower() for c in s if c.isalnum())
    return cleaned == cleaned[::-1]

def is_anagram(s1, s2):
    return Counter(s1.lower()) == Counter(s2.lower())

def most_frequent(s):
    return Counter(s).most_common(1)[0]

def compress(s):
    if not s:
        return s
    result, count = [], 1
    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            count += 1
        else:
            result.append(s[i-1] + str(count))
            count = 1
    result.append(s[-1] + str(count))
    compressed = "".join(result)
    return compressed if len(compressed) < len(s) else s

print("=== Reverse ===")
print(f"  reverse('hello') = {reverse_string('hello')!r}")
print()
print("=== Palindrome ===")
for t in ["racecar", "A man, a plan, a canal: Panama", "hello"]:
    print(f"  {t!r:45s} -> {is_palindrome(t)}")
print()
print("=== Anagram ===")
for a, b in [("listen", "silent"), ("hello", "world"), ("Astronomer", "Moon starer")]:
    print(f"  is_anagram({a!r}, {b!r}) = {is_anagram(a, b)}")
print()
print("=== Most Frequent Char ===")
for s in ["hello world", "mississippi"]:
    char, count = most_frequent(s)
    print(f"  {s!r:20s} -> {char!r} ({count} times)")
print()
print("=== Run-Length Encoding ===")
for s in ["aaabbc", "aabbccdd", "abcd", "aaaaaaa"]:
    print(f"  compress({s!r}) = {compress(s)!r}")
```

## 9. Interview Q Demos

```python
import timeit
print("Q1: String concatenation O(n^2) vs join O(n)")

def bad_concat(n):
    s = ""
    for i in range(n):
        s += str(i)
    return s

def good_join(n):
    return "".join(str(i) for i in range(n))

for n in [100, 1000, 5000]:
    t_bad  = timeit.timeit(lambda: bad_concat(n), number=200)
    t_good = timeit.timeit(lambda: good_join(n), number=200)
    print(f"  n={n:5d}: += {t_bad:.4f}s, join {t_good:.4f}s, speedup {t_bad/t_good:.1f}x")
```

```python
print("Q8: Palindrome two-pointer O(1) space")

def is_palindrome_twoptr(s):
    cleaned = [c.lower() for c in s if c.isalnum()]
    l, r = 0, len(cleaned) - 1
    while l < r:
        if cleaned[l] != cleaned[r]:
            return False
        l += 1
        r -= 1
    return True

tests = [
    "racecar",
    "A man, a plan, a canal: Panama",
    "Was it a car or a cat I saw?",
    "hello world",
]
for s in tests:
    print(f"  {s!r[:40]:42s} -> {is_palindrome_twoptr(s)}")
```
