## 1. Truthiness Rules

```python
# All falsy values in Python
falsy_values = [
    False, 0, 0.0, 0j, "", b"", [], (), {}, set(), frozenset(), None
]

print("Falsy values:")
for v in falsy_values:
    print(f"  bool({repr(v):15s}) = {bool(v)}")

print()
print("Truthy examples:")
truthy_values = [True, 1, -1, 0.001, "0", " ", [0], (False,), {0: 0}]
for v in truthy_values:
    print(f"  bool({repr(v):15s}) = {bool(v)}")
```

```python
# Short-circuit evaluation
print("=== Short-circuit: 'and' ===")
print(f"False and (1/0)  = {False and (1/0)}"  # 1/0 never evaluated
      if True else "")
# Demonstrate safely:
x = None
result = x and x.upper()  # x is falsy, x.upper() never called
print(f"None and None.upper() = {result}")

print()
print("=== Short-circuit: 'or' ===")
name = ""
display = name or "Anonymous"
print(f"'' or 'Anonymous' = {display!r}")

name = "Alice"
display = name or "Anonymous"
print(f"'Alice' or 'Anonymous' = {display!r}")
```

```python
# Walrus operator (Python 3.8+)
import re

texts = ["Order #12345", "No order here", "Invoice #99"]

print("Walrus operator demo:")
for text in texts:
    if m := re.search(r'#(\d+)', text):
        print(f"  Found number: {m.group(1)} in '{text}'")
    else:
        print(f"  No number in '{text}'")
```

## 2. range() Internals

```python
import sys

print("=== range() is memory-efficient ===")
for n in [10, 1_000, 1_000_000, 1_000_000_000]:
    r = range(n)
    print(f"range({n:>12,}): {sys.getsizeof(r)} bytes")

print()
print("Compare with list:")
for n in [10, 1_000, 10_000]:
    lst = list(range(n))
    print(f"list(range({n:>6,})): {sys.getsizeof(lst):>8,} bytes")

print()
print("=== range() supports O(1) operations ===")
r = range(1_000_000)
print(f"999_999 in range(1_000_000): {999_999 in r}  (O(1) arithmetic check)")
print(f"range(1_000_000)[-1]: {r[-1]}")
print(f"len(range(1_000_000)): {len(r)}")
```

## 3. enumerate() and zip()

```python
fruits = ['apple', 'banana', 'cherry']
prices = [1.20, 0.50, 2.00]

print("=== enumerate() ===")
for i, fruit in enumerate(fruits, start=1):
    print(f"  {i}. {fruit}")

print()
print("=== zip() ===")
for fruit, price in zip(fruits, prices):
    print(f"  {fruit}: ${price:.2f}")

print()
print("=== zip_longest() ===")
from itertools import zip_longest
a = [1, 2, 3]
b = [10, 20]
for x, y in zip_longest(a, b, fillvalue=0):
    print(f"  {x} + {y} = {x + y}")

print()
print("=== zip to create dict ===")
keys = ['name', 'age', 'city']
vals = ['Alice', 30, 'NYC']
d = dict(zip(keys, vals))
print(f"  {d}")
```

## 4. for/while else Clause

```python
def find_prime_factor(n):
    """Use for/else to detect primes."""
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print(f"  {n} = {i} × {n//i}")
            break
    else:
        print(f"  {n} is prime")

print("Prime factor check:")
for n in [2, 7, 15, 17, 100, 97]:
    find_prime_factor(n)

print()
print("=== while/else ===")
import random
random.seed(42)
attempts = 0
secret = 7
guesses = [3, 9, 7]  # simulated guesses

for guess in guesses:
    attempts += 1
    if guess == secret:
        print(f"  Found {secret} in {attempts} attempts!")
        break
else:
    print(f"  {secret} not found in {attempts} attempts")
```

## 6. Comprehensions

```python
import sys

print("=== List Comprehension ===")
squares = [x**2 for x in range(10) if x % 2 == 0]
print(f"Even squares: {squares}")

# Nested
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flat = [x for row in matrix for x in row]
print(f"Flattened: {flat}")

print()
print("=== Dict Comprehension ===")
words = ["hello", "world", "python"]
word_len = {w: len(w) for w in words}
print(f"Word lengths: {word_len}")

# Invert a dict
inverted = {v: k for k, v in word_len.items()}
print(f"Inverted: {inverted}")

print()
print("=== Set Comprehension ===")
unique_lengths = {len(w) for w in words}
print(f"Unique lengths: {unique_lengths}")

print()
print("=== Generator Expression (memory efficient) ===")
gen = (x**2 for x in range(1000))
lst = [x**2 for x in range(1000)]
print(f"Generator size: {sys.getsizeof(gen)} bytes")
print(f"List size:      {sys.getsizeof(lst)} bytes")
print(f"Sum via generator: {sum(x**2 for x in range(1000))}")
```

## 7. Performance Benchmarks

```python
import timeit

N = 100_000
SIZE = 1000

# Benchmark 1: List comprehension vs append loop
def loop_append():
    result = []
    for x in range(SIZE):
        result.append(x**2)
    return result

def list_comp():
    return [x**2 for x in range(SIZE)]

t_loop = timeit.timeit(loop_append, number=N)
t_comp = timeit.timeit(list_comp, number=N)

print(f"Loop + append:      {t_loop:.3f}s")
print(f"List comprehension: {t_comp:.3f}s")
print(f"Comprehension is {t_loop/t_comp:.2f}x faster")
print()
```

```python
import timeit

# Benchmark 2: Attribute caching in loops
def without_cache():
    result = []
    for i in range(1000):
        result.append(i)  # 'append' looked up each time
    return result

def with_cache():
    result = []
    _append = result.append  # cache the method
    for i in range(1000):
        _append(i)
    return result

t1 = timeit.timeit(without_cache, number=50_000)
t2 = timeit.timeit(with_cache, number=50_000)

print(f"Without method cache: {t1:.3f}s")
print(f"With method cache:    {t2:.3f}s")
print(f"Speedup: {t1/t2:.2f}x")
print()

# Benchmark 3: Built-in sum vs manual loop
data = list(range(10000))

def manual_sum(data):
    total = 0
    for x in data:
        total += x
    return total

t_manual = timeit.timeit(lambda: manual_sum(data), number=10_000)
t_builtin = timeit.timeit(lambda: sum(data), number=10_000)

print(f"Manual sum loop: {t_manual:.3f}s")
print(f"Built-in sum():  {t_builtin:.3f}s")
print(f"Built-in is {t_manual/t_builtin:.1f}x faster")
```

## 8. break / continue / pass Demo

```python
print("=== break ===")
for i in range(10):
    if i == 5:
        print(f"  Breaking at i={i}")
        break
    print(f"  i={i}")

print()
print("=== continue ===")
for i in range(10):
    if i % 2 == 0:
        continue  # skip even numbers
    print(f"  i={i} (odd)")

print()
print("=== pass ===")
for i in range(5):
    if i == 2:
        pass  # placeholder — does nothing
    print(f"  i={i}")
```

## 9. Interview Q Demos

```python
# Q2: for/else — search without flag variable
print("Q2: for/else pattern")

def search_with_flag(lst, target):
    found = False
    for item in lst:
        if item == target:
            found = True
            break
    if not found:
        return "not found"
    return "found"

def search_with_else(lst, target):
    for item in lst:
        if item == target:
            return "found"
    else:
        return "not found"

data = [1, 3, 5, 7, 9]
print(f"  search_with_flag([1,3,5,7,9], 5) = {search_with_flag(data, 5)}")
print(f"  search_with_else([1,3,5,7,9], 5) = {search_with_else(data, 5)}")
print(f"  search_with_else([1,3,5,7,9], 4) = {search_with_else(data, 4)}")
```

```python
# Q3: range() O(1) membership
import timeit
print("Q3: range() O(1) membership vs list O(n)")

N = 10_000_000
r = range(N)
lst = list(range(N))

t_range = timeit.timeit(f"{N-1} in r", globals={'r': r}, number=100_000)
t_list  = timeit.timeit(f"{N-1} in lst", globals={'lst': lst}, number=1000)

print(f"  range membership (100k lookups): {t_range*1000:.2f} ms")
print(f"  list  membership (1k lookups):   {t_list*1000:.2f} ms")
print(f"  range is ~{(t_list/1000)/(t_range/100000):.0f}x faster per lookup")
```
