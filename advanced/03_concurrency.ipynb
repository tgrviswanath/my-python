## 1. GIL Demonstration

```python
import sys
import threading
import time

print(f'GIL switch interval: {sys.getswitchinterval()*1000:.1f}ms')

def cpu_task(n):
    """CPU-bound: count down from n."""
    while n > 0:
        n -= 1

N = 5_000_000

# Single thread
start = time.perf_counter()
cpu_task(N)
cpu_task(N)
single_time = time.perf_counter() - start

# Two threads (GIL prevents true parallelism for CPU-bound)
t1 = threading.Thread(target=cpu_task, args=(N,))
t2 = threading.Thread(target=cpu_task, args=(N,))
start = time.perf_counter()
t1.start(); t2.start()
t1.join(); t2.join()
thread_time = time.perf_counter() - start

print(f'Single thread: {single_time:.3f}s')
print(f'Two threads:   {thread_time:.3f}s')
print(f'Speedup: {single_time/thread_time:.2f}x (expect ~1.0 due to GIL)')
```

## 2. Race Condition Demo

```python
import threading

# Unsafe counter
unsafe_counter = 0

def unsafe_increment(n):
    global unsafe_counter
    for _ in range(n):
        unsafe_counter += 1  # NOT atomic: LOAD, ADD, STORE

# Safe counter with lock
safe_counter = 0
lock = threading.Lock()

def safe_increment(n):
    global safe_counter
    for _ in range(n):
        with lock:
            safe_counter += 1

N = 50_000
THREADS = 4

# Run unsafe
unsafe_counter = 0
threads = [threading.Thread(target=unsafe_increment, args=(N,)) for _ in range(THREADS)]
for t in threads: t.start()
for t in threads: t.join()
print(f'Unsafe counter: {unsafe_counter} (expected {N*THREADS})')
print(f'Lost updates: {N*THREADS - unsafe_counter}')

# Run safe
safe_counter = 0
threads = [threading.Thread(target=safe_increment, args=(N,)) for _ in range(THREADS)]
for t in threads: t.start()
for t in threads: t.join()
print(f'Safe counter: {safe_counter} (expected {N*THREADS})')
```

## 3. Threading: Producer-Consumer with Queue

```python
import queue
import threading
import time
import random

def producer(q, n, name):
    for i in range(n):
        item = f'{name}-item-{i}'
        q.put(item)
        time.sleep(random.uniform(0.01, 0.05))
    q.put(None)  # sentinel

def consumer(q, results, name):
    while True:
        item = q.get()
        if item is None:
            q.put(None)  # pass sentinel to next consumer
            break
        results.append(f'{name} processed {item}')
        q.task_done()

q = queue.Queue(maxsize=10)
results = []

p = threading.Thread(target=producer, args=(q, 5, 'P1'))
c1 = threading.Thread(target=consumer, args=(q, results, 'C1'))
c2 = threading.Thread(target=consumer, args=(q, results, 'C2'))

p.start(); c1.start(); c2.start()
p.join(); c1.join(); c2.join()

print(f'Processed {len(results)} items:')
for r in sorted(results):
    print(f'  {r}')
```

## 4. ThreadPoolExecutor  I/O Bound Benchmark

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

def simulate_io(task_id, delay):
    """Simulate I/O-bound work (e.g., network request)."""
    time.sleep(delay)
    return f'Task {task_id} done after {delay:.2f}s'

delays = [0.3, 0.1, 0.2, 0.15, 0.25, 0.05, 0.35, 0.1]

# Sequential
start = time.perf_counter()
sequential_results = [simulate_io(i, d) for i, d in enumerate(delays)]
sequential_time = time.perf_counter() - start

# Concurrent with ThreadPoolExecutor
start = time.perf_counter()
with ThreadPoolExecutor(max_workers=4) as executor:
    futures = {executor.submit(simulate_io, i, d): i for i, d in enumerate(delays)}
    concurrent_results = []
    for future in as_completed(futures):
        concurrent_results.append(future.result())
concurrent_time = time.perf_counter() - start

print(f'Sequential time: {sequential_time:.3f}s')
print(f'Concurrent time: {concurrent_time:.3f}s')
print(f'Speedup: {sequential_time/concurrent_time:.2f}x')
print('Results (in completion order):')
for r in concurrent_results:
    print(f'  {r}')
```

## 5. asyncio  Concurrent Coroutines

```python
import asyncio

async def async_fetch(url, delay):
    """Simulate async HTTP request."""
    print(f'  Fetching {url}...')
    await asyncio.sleep(delay)  # non-blocking wait
    return f'Response from {url} ({delay:.2f}s)'

async def main_gather():
    urls = [
        ('http://api1.example.com', 0.3),
        ('http://api2.example.com', 0.1),
        ('http://api3.example.com', 0.2),
        ('http://api4.example.com', 0.15),
    ]
    start = asyncio.get_event_loop().time()
    results = await asyncio.gather(*[async_fetch(url, delay) for url, delay in urls])
    elapsed = asyncio.get_event_loop().time() - start
    print(f'All done in {elapsed:.3f}s (max delay was 0.3s)')
    for r in results:
        print(f'  {r}')

await main_gather()
```

```python
# asyncio.Queue  producer/consumer
async def async_producer(queue, items):
    for item in items:
        await queue.put(item)
        await asyncio.sleep(0.05)
    await queue.put(None)  # sentinel

async def async_consumer(queue, name, results):
    while True:
        item = await queue.get()
        if item is None:
            await queue.put(None)  # pass sentinel
            break
        results.append(f'{name}: {item}')
        queue.task_done()

async def main_queue():
    q = asyncio.Queue(maxsize=5)
    results = []
    await asyncio.gather(
        async_producer(q, range(6)),
        async_consumer(q, 'Consumer-A', results),
        async_consumer(q, 'Consumer-B', results),
    )
    for r in results:
        print(r)

await main_queue()
```

## 6. asyncio Timeout & Error Handling

```python
async def slow_operation(name, delay):
    await asyncio.sleep(delay)
    return f'{name} completed'

async def main_timeout():
    # wait_for with timeout
    try:
        result = await asyncio.wait_for(slow_operation('fast', 0.1), timeout=0.5)
        print(f'Success: {result}')
    except asyncio.TimeoutError:
        print('Timed out!')

    try:
        result = await asyncio.wait_for(slow_operation('slow', 2.0), timeout=0.5)
        print(f'Success: {result}')
    except asyncio.TimeoutError:
        print('Timed out as expected!')

    # gather with return_exceptions
    async def might_fail(n):
        if n == 2:
            raise ValueError(f'Task {n} failed')
        return f'Task {n} ok'

    results = await asyncio.gather(
        *[might_fail(i) for i in range(4)],
        return_exceptions=True
    )
    for i, r in enumerate(results):
        if isinstance(r, Exception):
            print(f'Task {i} error: {r}')
        else:
            print(f'Task {i} result: {r}')

await main_timeout()
```

## 7. Semaphore  Limit Concurrent Access

```python
async def limited_fetch(semaphore, url, delay):
    async with semaphore:
        print(f'  Fetching {url} (semaphore acquired)')
        await asyncio.sleep(delay)
        return f'Done: {url}'

async def main_semaphore():
    # Allow max 2 concurrent requests
    semaphore = asyncio.Semaphore(2)
    urls = [(f'url{i}', 0.2) for i in range(6)]
    start = asyncio.get_event_loop().time()
    results = await asyncio.gather(*[limited_fetch(semaphore, url, d) for url, d in urls])
    elapsed = asyncio.get_event_loop().time() - start
    print(f'Completed {len(results)} requests in {elapsed:.2f}s')
    print('(With semaphore=2, 6 requests at 0.2s each = ~0.6s in 3 batches)')

await main_semaphore()
```

## 8. Threading vs asyncio Benchmark (I/O-bound)

```python
import time
from concurrent.futures import ThreadPoolExecutor

N_TASKS = 20
DELAY = 0.1

# Threading approach
def blocking_io(i):
    time.sleep(DELAY)
    return i

start = time.perf_counter()
with ThreadPoolExecutor(max_workers=N_TASKS) as ex:
    thread_results = list(ex.map(blocking_io, range(N_TASKS)))
thread_time = time.perf_counter() - start

# asyncio approach
async def async_io(i):
    await asyncio.sleep(DELAY)
    return i

async def run_async():
    return await asyncio.gather(*[async_io(i) for i in range(N_TASKS)])

start = time.perf_counter()
async_results = await run_async()
async_time = time.perf_counter() - start

print(f'{N_TASKS} tasks, each {DELAY}s I/O:')
print(f'  Threading: {thread_time:.3f}s')
print(f'  asyncio:   {async_time:.3f}s')
print(f'  Sequential would be: {N_TASKS * DELAY:.1f}s')
```
