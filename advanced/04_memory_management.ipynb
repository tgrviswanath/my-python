## 1. Reference Counting with sys.getrefcount

```python
import sys

x = [1, 2, 3]
print(f'After x = [1,2,3]:          refcount = {sys.getrefcount(x)} (x + getrefcount arg)')

y = x
print(f'After y = x:                refcount = {sys.getrefcount(x)} (x, y + arg)')

lst = [x, x, x]
print(f'After lst = [x,x,x]:        refcount = {sys.getrefcount(x)} (x, y, 3 in lst + arg)')

del lst
print(f'After del lst:              refcount = {sys.getrefcount(x)}')

del y
print(f'After del y:                refcount = {sys.getrefcount(x)}')

# Integer caching: -5 to 256 are singletons
a, b = 256, 256
print(f'\n256 is 256: {a is b}  (cached singleton)')
a, b = 257, 257
print(f'257 is 257: {a is b}  (not cached, usually False)')
```

## 2. Cyclic Garbage Collector

```python
import gc

print('GC thresholds:', gc.get_threshold())
print('GC counts:', gc.get_count())

# Create a reference cycle
class Node:
    def __init__(self, name):
        self.name = name
        self.ref = None
    def __repr__(self):
        return f'Node({self.name})'

# Disable GC to observe cycle accumulation
gc.disable()
gc.collect()  # clean slate

# Create cycles
cycles_created = 0
for _ in range(100):
    a = Node('a')
    b = Node('b')
    a.ref = b
    b.ref = a  # cycle!
    del a, b
    cycles_created += 1

print(f'\nCreated {cycles_created} cycles, GC disabled')
print('Objects tracked by GC:', len(gc.get_objects()))

# Now collect
collected = gc.collect()
print(f'GC collected: {collected} objects')
gc.enable()
```

## 3. tracemalloc  Memory Profiling

```python
import tracemalloc

tracemalloc.start()

# Allocate some memory
data1 = [i**2 for i in range(10_000)]
data2 = {str(i): i for i in range(5_000)}
data3 = 'x' * 100_000

snapshot = tracemalloc.take_snapshot()
top_stats = snapshot.statistics('lineno')

print('Top 5 memory allocations:')
for stat in top_stats[:5]:
    print(f'  {stat}')

current, peak = tracemalloc.get_traced_memory()
print(f'\nCurrent memory: {current / 1024:.1f} KB')
print(f'Peak memory:    {peak / 1024:.1f} KB')

tracemalloc.stop()
```

```python
# Compare two snapshots to find memory growth
tracemalloc.start()
snapshot1 = tracemalloc.take_snapshot()

# Simulate a leaky operation
_leak_cache = {}
for i in range(1000):
    _leak_cache[i] = [j**2 for j in range(100)]

snapshot2 = tracemalloc.take_snapshot()
stats = snapshot2.compare_to(snapshot1, 'lineno')

print('Memory growth (top 5):')
for stat in stats[:5]:
    print(f'  {stat}')

tracemalloc.stop()
```

## 4. __slots__ Memory Optimization

```python
import sys
import tracemalloc

class WithDict:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

class WithSlots:
    __slots__ = ('x', 'y', 'z')
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

# Single instance comparison
d = WithDict(1, 2, 3)
s = WithSlots(1, 2, 3)

dict_size = sys.getsizeof(d) + sys.getsizeof(d.__dict__)
slot_size = sys.getsizeof(s)

print(f'WithDict instance: {sys.getsizeof(d)} bytes')
print(f'WithDict __dict__: {sys.getsizeof(d.__dict__)} bytes')
print(f'WithDict total:    {dict_size} bytes')
print(f'WithSlots total:   {slot_size} bytes')
print(f'Memory savings:    {(1 - slot_size/dict_size)*100:.1f}%')

# Bulk comparison
N = 100_000

tracemalloc.start()
dict_objects = [WithDict(i, i*2, i*3) for i in range(N)]
dict_peak = tracemalloc.get_traced_memory()[1]
tracemalloc.stop()

tracemalloc.start()
slot_objects = [WithSlots(i, i*2, i*3) for i in range(N)]
slot_peak = tracemalloc.get_traced_memory()[1]
tracemalloc.stop()

print(f'\n{N:,} instances:')
print(f'  WithDict peak:  {dict_peak / 1024 / 1024:.2f} MB')
print(f'  WithSlots peak: {slot_peak / 1024 / 1024:.2f} MB')
print(f'  Savings: {(1 - slot_peak/dict_peak)*100:.1f}%')
```

## 5. Weak References

```python
import weakref
import gc

class ExpensiveObject:
    def __init__(self, name):
        self.name = name
        print(f'  Created: {name}')
    def __del__(self):
        print(f'  Deleted: {self.name}')

print('--- Strong reference ---')
obj = ExpensiveObject('obj1')
strong_ref = obj
del obj
print('After del obj (strong_ref still holds it)')
del strong_ref
print('After del strong_ref')

print('\n--- Weak reference ---')
obj = ExpensiveObject('obj2')
weak = weakref.ref(obj)
print(f'Weak ref alive: {weak() is not None}')
del obj
gc.collect()
print(f'Weak ref alive after del: {weak() is not None}')

print('\n--- WeakValueDictionary cache ---')
cache = weakref.WeakValueDictionary()
obj = ExpensiveObject('cached')
cache['key'] = obj
print(f'Cache size: {len(cache)}')
del obj
gc.collect()
print(f'Cache size after del: {len(cache)}')  # 0  auto-removed
```

## 6. Memory Leak Detection

```python
import tracemalloc
import gc

# Simulate a memory leak: global cache without eviction
_global_cache = {}

def leaky_process(key):
    """Leaks memory: cache grows without bound."""
    _global_cache[key] = [i**2 for i in range(1000)]
    return _global_cache[key]

def fixed_process(key, max_size=100):
    """Fixed: evict oldest entries when cache is full."""
    if len(_global_cache) >= max_size:
        oldest = next(iter(_global_cache))
        del _global_cache[oldest]
    _global_cache[key] = [i**2 for i in range(1000)]
    return _global_cache[key]

# Detect the leak
tracemalloc.start()
snapshot1 = tracemalloc.take_snapshot()

for i in range(200):
    leaky_process(f'key_{i}')

snapshot2 = tracemalloc.take_snapshot()
stats = snapshot2.compare_to(snapshot1, 'lineno')

print('Memory growth from leaky_process:')
for stat in stats[:3]:
    print(f'  {stat}')

# Fix: clear and use bounded cache
_global_cache.clear()
snapshot3 = tracemalloc.take_snapshot()

for i in range(200):
    fixed_process(f'key_{i}')

snapshot4 = tracemalloc.take_snapshot()
stats2 = snapshot4.compare_to(snapshot3, 'lineno')

print('\nMemory growth from fixed_process (bounded cache):')
for stat in stats2[:3]:
    print(f'  {stat}')

tracemalloc.stop()
```

## 7. Copy Semantics

```python
import copy

original = [[1, 2, 3], [4, 5, 6], {'key': 'value'}]

# Assignment  same object
assigned = original
assigned[0].append(99)
print(f'After assignment mutation, original[0]: {original[0]}')

# Reset
original = [[1, 2, 3], [4, 5, 6], {'key': 'value'}]

# Shallow copy
shallow = copy.copy(original)
shallow.append('new item')  # doesn't affect original
shallow[0].append(99)       # DOES affect original (shared nested list)
print(f'After shallow copy mutation:')
print(f'  original: {original}')
print(f'  shallow:  {shallow}')

# Reset
original = [[1, 2, 3], [4, 5, 6], {'key': 'value'}]

# Deep copy
deep = copy.deepcopy(original)
deep[0].append(99)
deep[2]['key'] = 'modified'
print(f'After deep copy mutation:')
print(f'  original: {original}')
print(f'  deep:     {deep}')
```

```python
import timeit

nested = [list(range(100)) for _ in range(100)]

t_shallow = timeit.timeit(lambda: copy.copy(nested), number=10_000)
t_deep = timeit.timeit(lambda: copy.deepcopy(nested), number=1_000)

print(f'Shallow copy (10k runs): {t_shallow:.3f}s')
print(f'Deep copy (1k runs):     {t_deep:.3f}s')
print(f'Deep copy is ~{t_deep/t_shallow*10:.0f}x slower per call')
```

## 8. sys.getsizeof  Object Sizes

```python
import sys

objects = [
    ('None', None),
    ('True', True),
    ('int 0', 0),
    ('int 1000', 1000),
    ('float', 3.14),
    ('str empty', ''),
    ('str hello', 'hello'),
    ('list empty', []),
    ('list [1,2,3]', [1,2,3]),
    ('tuple (1,2,3)', (1,2,3)),
    ('dict empty', {}),
    ('dict {a:1}', {'a':1}),
    ('set empty', set()),
    ('set {1,2,3}', {1,2,3}),
    ('bytes 10', b'x'*10),
    ('bytearray 10', bytearray(10)),
]

print(f'{"Object":<20} {"Size (bytes)":>12}')
print('-' * 34)
for name, obj in objects:
    print(f'{name:<20} {sys.getsizeof(obj):>12,}')
```
